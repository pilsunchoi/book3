# 시작하며

이 책 《**데이터 기반 사회과학 분석: 교육과 소득분배**》는 데이터 분석과 사회과학 연구를 결합하여 실제 문제를 실증적으로 탐구하고 해결하고자 하는 독자들을 위한 길잡이입니다. 다양한 연구 주제를 중심으로 파이썬의 강력한 분석 도구를 활용하는 방법을 제시하며, 데이터의 의미를 발견하고 통찰을 얻는 과정을 체계적으로 안내합니다.

이 책은 각 장이 하나의 독립된 연구 프로젝트처럼 구성되어 있습니다. 독자들은 데이터의 수집 정리 분석, 그리고 결과의 해석까지의 과정을 단계별로 따라가며 파이썬을 도구로 삼아 주어진 연구 문제를 해결하는 방법을 익힐 수 있습니다. 

- **다문화청소년의 학업성취도 분석**  
  《다문화청소년패널조사》를 패널 데이터 형식으로 구축하고 패널 회귀분석 기법을 이용해 다문화청소년들의 학업 성취도에 영향을 미치는 요인들을 탐색합니다.   

- **다문화 배경과 학업성적**  
  《한국아동·청소년패널조사》를 패널 데이터 형식으로 구축하고 패널 회귀분석 기법을 이용해 다문화 배경이 학생들의 학업성적에 미치는 영향을 분석합니다.  

- **청년 세대의 자산불평등 추정**  
  《가계금융복지조사》를 활용하여 지니계수를 계산하고, 2030세대와 청년 세대의 자산불평등을 측정하고 2010년 이후 변화 추이와 그 요인을 탐색합니다.  

- **순자산 분포 추정**  
  파라미터형 분포 함수를 사용해 순자산(net worth)의 분포를 최대가능도추정(MLE: Maximum Likelihood Estimation) 기법으로 추정하는 방법을 살펴봅니다.  

- **나이별 순자산 분포 추정**  
  파라미터형 이변량 분포 함수를 사용해 나이와 자산의 결합분포(joint distribution)를 MLE 기법으로 추정하는 방법을 살펴봅니다.

파이썬은 방대한 라이브러리와 직관적인 코드 구조 덕분에 연구와 데이터 분석을 위한 강력한 도구로 자리 잡았습니다. 또한 ChatGPT와 같은 생성형 GPT와 함께 사용할 경우 데이터 분석에서 시너지 효과를 얻을 수 있습니다. 본 책은 파이썬 초보자부터 중급자까지 누구나 쉽게 따라 할 수 있도록 작성됐습니다. 저자들의 또다른 책 《[**데이터과학의 첫걸음: 빅데이터 분석 with Python**](https://pilsunchoi.github.io/book1/notebooks/00_서문.html)》을 함께 활용하면 더 효율적인 학습이 가능할 것입니다. 이 책에서 사용한 데이터세트는 [https://github.com/pilsunchoi/book3/tree/main/Data](https://github.com/pilsunchoi/book3/tree/main/Data)에서 다운로드 받을 수 있습니다.

이 책은 사회과학 연구자, 데이터 분석가, 또는 연구에 관심을 가진 학생들에게 유용할 것입니다. 여러분이 실제 데이터를 다루고, 그 안에서 의미를 찾아내며, 세상을 이해하는 새로운 방법을 배울 수 있기를 바랍니다.
