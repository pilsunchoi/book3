## 1 서론

본 연구는 부, 소득, 소비 및 기타 경제적 웰빙 변수 간의 결합분포(joint distribution)를 추정하기 위해 Johnson(1949b)의 다변량(multivariate) $S_U$ 분포를 사용하는 것을 제안한다. 이 분포는 정규분포를 변환하여 만들어지기 때문에 다변량 분포를 간단하게 생성시킬 수 있다. 그 한계분포(marginal distribution)는 매우 높은 유연성을 지니며, 넓은 범위의 왜도와 첨도를 포착할 수 있다. 특히, $S_U$ 분포는 순자산이나 가처분 소득과 같이 음의 값이 포함돼있는 변수에도 적용 가능하다. 이러한 장점으로 인해 Choi and Min(2025)은 순자산과 소득에 단변량 $S_U$ 분포를 도입하여 적합성을 입증하였다. 본 연구에서는 이를 다변량 차원으로 확장하고자 한다.

소득과 부와 같은 두 경제적 웰빙 변수 간의 상관관계를 측정하기 위해 Pearson 상관계수 대신 "정규화된(normalized)" 상관계수의 사용을 권한다. 이는 이러한 변수들이 일반적으로 높은 왜도를 가지며 극단적인 이상값을 포함하고 있어 변수 간의 관계를 왜곡시킬 수 있기 때문이다. 정규화된 상관계수는 각 한계분포를 정규분포로 변환한 후, 이를 바탕으로 Pearson 상관계수를 측정함으로써 도출된다. 이는 사실상 이변량 $S_U$ 분포를 정의하는 파라미터 중 하나로 추정된다. 

사례 연구로, 미국 서베이 데이터를 사용하여 순자산과 나이의 결합분포를 추정하는 데 이변량 $S_U$ 모델을 적용하였다.

## 2 Johnson SU 분포

$S_U$ 분포는 Johnson(1949a)의 선구적인 연구에서 처음 등장했다. $S_U$ 변수 $X$는 아래와 같은 방식으로 정규분포의 변환을 통해 생성된다.

$$
\sinh^{-1}\left(\frac{X - m}{s}\right) = \lambda + \theta Z, \quad -\infty < X < \infty, \ s > 0, \ \theta > 0 \tag{1}
$$

여기서 $Z$는 표준정규(standard normal) 변수이다. $S_U$는 "unbounded system"을 의미하며, $X$의 범위가 제한되지 않음을 나타낸다. $S_U$ 분포의 확률밀도함수(PDF)는 다음과 같다.

$$
f(x) = J \cdot \phi(z), \tag{2}
$$

여기서 $J = \theta^{-1} \left[(x - m)^2 + s^2\right]^{-1/2}$, $z = \theta^{-1} \left[\sinh^{-1}\left(\frac{x - m}{s}\right) - \lambda\right]$, $\phi(\cdot)$는 표준 정규 변수의 PDF이다. $X$의 평균과 분산은 각각 $m + s \cdot e^{\theta^2 / 2} \sinh(\lambda)$와 $\frac{1}{2} s^2 \left(e^{\theta^2} - 1\right)\left(e^{\theta^2} \cosh(2\lambda) + 1\right)$이다.

식 1의 네 파라미터 중 $\lambda$의 부호는 왜도의 방향을 결정한다. $\lambda = 0$일 때 분포는 대칭이며, $\lambda > 0$($\lambda < 0$)은 양의(음의) 왜도를 나타낸다. $\lambda$와 $\theta$가 0에서 멀어질수록 분포는 정규분포와 더 멀어지며, 비대칭성과 분포의 꼬리 두께가 증가한다. Johnson(1949a)은 $S_U$ 분포가 매우 유연한 분포로, 왜도와 첨도의 광범위한 조합을 포착할 수 있음을 보였다. 또한, 로그정규 분포를 비롯해 기존의 소득 분포 추정에 사용되는 파라미터형 분포와 달리, $S_U$ 분포는 순자산이나 가처분소득과 같은 음의 값이 포함된 변수에도 적용 가능하다.

$S_U$ 분포는 다변량 차원으로 쉽게 확장 가능하다(Johnson, 1949b). $N \times 1$ 확률변수 벡터 $Z$가 다변량 표준정규분포를 따를 때, $Z$의 결합(joint) PDF는 다음과 같다.

$$
\phi_R(z) = (2\pi)^{-N/2} |R|^{-1/2} \exp\left(-\frac{1}{2} z' R^{-1} z\right), \tag{3}
$$

여기서 $R$은 상관계수 행렬이며, $|R|$은 $R$의 행렬식(determinant)이다. 다변량 $S_U$ 확률변수 벡터 $X$는 각 변수 $X_i$를 정규 변수로 변환하는 역 쌍곡선 사인 변환(inverse hyperbolic sine transformation)을 통해 생성된다. 즉, 다음과 같다.

$$
\sinh^{-1}\left(\frac{X_i - m_i}{s_i}\right) = \lambda_i + \theta_i Z_i, \quad s_i > 0, \ \theta_i > 0.
$$

따라서 $X$의 결합 PDF는 다음과 같다.

$$
f(x) = (2\pi)^{-N/2} |R|^{-1/2} J \cdot \exp\left(-\frac{1}{2} z' R^{-1} z\right), \tag{4}
$$

여기서 $z_i = \theta_i^{-1} \left[\sinh^{-1}\left(\frac{x_i - m_i}{s_i}\right) - \lambda_i\right]$와 $J = \prod_{i=1}^N \theta_i^{-1} \left[(x_i - m_i)^2 + s_i^2\right]^{-1/2}$이다.

식 4에서 $R$은 $X$에서 변환된 변수 $Z$의 상관행렬임에 유의해야 한다. 비선형 변환이기 때문에 $Z$의 상관관계는 $X$의 상관관계와 동일하지 않다. $X_i$와 $X_j$ 사이의 Pearson 상관계수 $\rho_{ij}$는 다음과 같다.

$$
\rho_{ij} = \frac{e^{(\theta_i^2 + \theta_j^2) / 2}}{\sigma_i \sigma_j} \left[\frac{1}{2} e^{r_{ij} \theta_i \theta_j} \cosh(\lambda_i + \lambda_j) - \frac{1}{2} e^{-r_{ij} \theta_i \theta_j} \cosh(\lambda_i - \lambda_j) - \sinh(\lambda_i) \sinh(\lambda_j)\right], \tag{5}
$$

여기서 $\sigma_k = \left[\frac{1}{2} \left(e^{\theta_k^2} - 1\right)\left(e^{\theta_k^2} \cosh(2\lambda_k) + 1\right)\right]^{1/2}, \ k = i, j$이고, $i = j$일 경우 $\rho_{ij}$는 1이 된다. 

반대로 $X$가 상관행렬 $\Sigma$를 가지는 다변량 $S_U$ 분포를 따를 때, $Z_i$와 $Z_j$ 사이의 상관관계 $r_{ij}$는 다음과 같다.

$$
r_{ij} = \frac{1}{\theta_i \theta_j} \ln\left(\frac{B_{ij} + \sqrt{B_{ij}^2 + \cosh(\lambda_i + \lambda_j) \cosh(\lambda_i - \lambda_j)}}{\cosh(\lambda_i + \lambda_j)}\right), \tag{6}
$$

여기서 $B_{ij} = \rho_{ij} \sigma_i \sigma_j \exp\left(-\frac{1}{2} (\theta_i^2 + \theta_j^2)\right) + \sinh(\lambda_i) \sinh(\lambda_j)$이다.

다음 절에서는 2022년 미국의 소비자금융조사(SCF: Survey of Consumer Finances)를 사용하여 순자산과 나이의 결합분포를 추정한다. 이변량 $S_U$ 분포는 다음과 같은 PDF를 갖는다.

$$
f(x_1, x_2) = \frac{1}{2\pi\sqrt{1 - r}} J \cdot \exp\left(-\frac{1}{2(1 - r^2)}\left(z_1^2 - 2r z_1 z_2 + z_2^2\right)\right), \tag{7}
$$

여기서 $z_1 = \theta_1^{-1} \left[\sinh^{-1}\left(\frac{x_1 - m_1}{s_1}\right) - \lambda_1\right]$, $z_2 = \theta_2^{-1} \left[\sinh^{-1}\left(\frac{x_2 - m_2}{s_2}\right) - \lambda_2\right]$, $J = (\theta_1 \theta_2)^{-1} \left[(x_1 - m_1)^2 + s_1^2\right]^{-1/2} \left[(x_2 - m_2)^2 + s_2^2\right]^{-1/2}$이다.

조건부(conditional) 분포 $X_1 \mid X_2 = x_2$는 $X_1$과 동일한 $S_U$ 시스템을 따르지만, $\lambda_1$과 $\theta_1$이 각각 $\lambda_1^* = \lambda_1 + r\theta_1 \theta_2^{-1}\left[\sinh^{-1}\left(\frac{x_2 - m_2}{s_2}\right) - \lambda_2\right]$ 및 $\theta_1^* = \theta_1 \sqrt{1 - r^2}$로 대체된다(Kotz et al., 2000).

$$
X_1 \mid X_2 = x_2 \sim S_U(m_1, s_1, \lambda_1^*, \theta_1^*). \tag{8}
$$

## 3 순자산과 나이에 대한 결합분포 추정

2022 SCF 데이터에서 순자산(백만 달러, 명목)과 나이 두 가지 변수를 선택하였다. 추정은 식 7을 기반으로 최대가능도추정(maximum likelihood estimation )을 사용하여 수행되었다. 사실, 추정을 2단계로 수행할 수 있는데, 첫 번째 단계에서는 각 개별적인 한계분포를 추정하고, 두 번째 단계에서는 그들 간의 상관관계 파라미터를 추정하는 방법이다. 이러한 2단계 추정법은 많은 수의 변수를 다룰 때 유용하게 사용될 수 있다. 그러나 본 연구에서는 두 개의 변수만 다루기 때문에 모든 파라미터를 한 번에 추정하였다. 최대가능도추정법은 파이썬 `SciPy`의 `optimize` 모듈과 limited-memory BFGS(`L-BFGS-B`) 알고리즘으로 수행되었다. 본 연구에서는 모든 통계가 SCF에서 제공된 가중치를 사용하여 계산되었다.  

<p style="text-align: center;"><b>표 1. 이변량 SU 분포의 추정된 파라미터</b></p>

| 변수        | $m$       | $s$      | $\lambda$ | $\theta$ | $r$    |  
|:-----------:|:--------:|:--------:|:--------:|:--------:|:--------:|  
| 순자산 ($X_1$) | -0.005   | 0.015    | 3.021    | 2.077    | 0.345    |  
| 나이 ($X_2$)   | -2.301   | 123.780  | 0.426    | 0.129    |          |  

표 1은 이변량 $S_U$ 분포의 추정 파라미터를 제시한다. 이를 바탕으로 나이에 따른 순자산의 조건부 분포를 추정하였다. 나이 중에서 우리는 특히 **대학 졸업 연령**에 해당하는 22, 23, 24세를 선택하였다. 2022 SCF 데이터에서 25세 이하에 대한 관측은 매우 제한적이다. 표 2에서 볼 수 있듯이, 22세에서 24세까지의 나이를 합친 경우에도 관측이 72개에 불과하다(전체의 1.58%). 이 작은 표본크기(sample size) 때문에 평균과 표준편차의 통계값이 나이에 따라 크게 달라진다. 그러나 이러한 경험적(empirical) 결과와 달리, $S_U$ 모델을 기반으로 한 22세, 23세, 24세의 추정값은 서로 매우 유사하다. 이 점에서 $S_U$ 모델은 주어진 조건에 맞는 관측수가 많지 않을 때 조건부 분포의 강건한(robust) 추정값을 얻는 데 유리함을 보여준다.

|:-----:|:-----:|:----:|:-----:|:-----:|:--------:|:-----------:|:---------------:|

이제 전체 관측에 있어서 순자산과 나이 간의 상관관계를 분석해보자. 두 변수 간의 피어슨 상관계수 추정값은 0.071이다. 상관관계의 부호는 예상대로 플러스이지만, 그 크기는 매우 작다. 이 낮은 피어슨 상관계수는 두 변수 간의 연관성(association)이 부족해서가 아니라, 순자산 변수의 분포에 극단적인 왜도와 과도한 첨도가 있기 때문일 수 있다. 그림 1은 이를 잘 보여준다. 그림의 패널 (a)는 순자산과 나이의 원본(original) 변수를 사용하여 산점도를 만들고, 패널 (b)는 "정규화된(normalized)" 변수를 사용한다. 여기서 "정규화"는 변수를 왜도와 초과첨도(excess kurtosis)가 0인 정규분포로 변환하는 것을 의미한다. 

우리의 경우, 정규화는 $\sinh^{-1} \left( \frac{x_{1i} - \hat{m}_1}{\hat{s}_1} \right)$와 $\sinh^{-1} \left( \frac{x_{2i} - \hat{m}_2}{\hat{s}_2} \right)$를 의미하며, 이 두 정규화된 변수의 산점도가 그림 1의 패널 (b)에 나와있다. 패널 (b)에서의 선형 상관은 패널 (a)보다 강한 것을 알 수 있다. 표 1에서 보듯이, 이 정규화된 변수들에 대한 피어슨 상관계수 추정값은 $\hat{r} = 0.345$로, 원래 변수 간의 상관계수인 0.071보다 훨씬 크다. 

그림 1은 부와 소득 변수에서 자주 발견되는 극단적인 왜도와 초과첨도가 변수 간의 연관성을 왜곡할 수 있음을 보여준다. 또한 $S_U$ 모델에 의해 추정된 정규화된 변수들 간의 상관관계가 연관성을 측정하는 데 더 나은 지표임을 보여준다.

## 4 결론

문헌에는 가계 소득과 부의 결합분포를 구축하기 위해 코퓰라(copula) 함수를 사용하는 접근법이 있다(Jäntti et al., 2015 등). 본 연구에서 논의한 다변량 $S_U$ 모델도 코퓰라 개념을 통해 이해할 수 있다. 즉, 다변량 $S_U$ 분포는 각 한계(marginal) $S_U$ 변수를 가우시안(Gaussian) 코퓰라 함수로 결합시킨 분포로 볼 수 있다. $S_U$ 변수는 정규분포에서 변환되었기 때문에 이를 다시 정규분포로 변환하고 가우시안 코퓰라를 사용하여 결합시키는 것이 매우 간단하다. 

$S_U$ 분포는 정규분포에서 변환된 분포이기 때문에 여러 가지 장점이 있다. 결합 확률밀도함수(PDF)가 상대적으로 간단한 형태이어서 최대가능도추정법을 한 단계로도 쉽게 수행할 수 있다. 또한, 다변량 $S_U$ 무작위 숫자(random numbers)를 생성하는 것도 간단하여 다변량 차원에서의 시뮬레이션 분석에 유리하다. 본 예제에서는 순자산과 나이라는 두 가지 변수를 다루었지만, 두 개 이상의 변수, 예를 들어 부, 소득, 소비의 결합분포를 추정할 때 다변량 $S_U$ 모델은 매력적인 선택지가 될 수 있다.
