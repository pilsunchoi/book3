# 1.1 다문화청소년 학업성취도 요인 탐색

## 1 연구 방법

### 연구 대상

다문화청소년패널조사는 다문화청소년의 성장․발달의 기초자료를 구축하고 발달추이를 분석함으로써, 다문화청소년을 위한 과학적인 정책수립과 정책추진의 근거를 마련하는 것을 목적으로 하고 있다.(<a href="https://www.nypi.re.kr/archive/board?menuId=MENU00221">다문화청소년패널조사 개요</a>.) 다문화청소년에 대한 근거없는 부정적 인식도 문제이겠지만, 반대로 그들에 대한 일방적 지원 주장 역시 다문화청소년들이 한국가정 출신의 청소년들에 비해 발달적 측면에서 부족하고 결핍된 존재로 인식하게 만들 수 있다는 점에서 바람직하지 않다. 다문화청소년의 발달 특성에 대한 보다 객관적이고 면밀한 분석 없이 이들에 대한 사회적 고정관념이 강화되는 현상이 지속될 경우 한국사회의 사회적 통합에 저해요소로 작용될 가능성이 있다는 문제의식으로부터 다문화청소년패널조사가 시작되었다. 

한국청소년정책연구원(NYPI)은 2011년 당시 초등학교 4학년 다문화청소년과 그들의 부모를 대상으로 패널을 구축하여 2020년까지 총 10개년(조사 주기: 1년)에 걸친 종단조사를 실시하였다. 또한 1기 패널들이 대학 진학을 앞둔 2019년부터는 기존의 1기 조사와는 별개로 새로운 패널을 구축하여 2기 종단연구를 진행하고 있다. 이들 1기 및 2기 다문화청소년패널조사에서는 다문화가정 자녀의 배경특성, 학교생활, 심리사회적응, 신체발달, 부모자녀관계, 정책지원 등 다양한 영역에 대한 종합적 조사가 이루어지고 있다. 1기 패널이 대학에 진학하기 시작한 2020년 조사에서는 이전 9차례의 조사와는 달리 다문화청소년의 현재 다양한 진로 상태에 초점을 맞춰 조사를 실시했으며, 학부모 조사는 대폭 축소하여 진행되었다. 1기 패널의 경우, 2011년 초등학교 4학년 학생과 그들의 학부모를 대상으로 총 1,625가구가 구축되었으며, 2021년 기준으로 최종 1,014가구가 유지되어 2011년 대비 62.4%의 유지율을 보이고 있다.(양계민 외, 2022)

본 연구는 1기 다문화청소년패널조사 데이터를 사용한다. 우리는 이 데이터를 이용해 다문화 학생들의 학교 성적이 그들의 성공적인 진로에서 얼마나 중요한 요소인지를 확인해보는 한편, 어떤 요인들이 학업성적에 영향을 미치는지, 그리고 이러한 관계를 분석하는 데 어떤 추정 모형이 적절한지 탐색한다.

## 2 추정 변수 및 모형

### 학업성적 변화 추이

다문화청소년패널조사에서 우리가 분석하고자 하는 항목은 “학교성적”이다. 이 항목은 청소년 본인이 스스로 평가하는 것으로 국어, 영어, 수학, 사회, 과학 등 5개 과목에 대해 1(매우 못하는 편이다)부터 5(매우 잘하는 편이다)까지 5점 척도로 평가됐다. 학교성적에 대한 평가는 1차조사부터 9차조사까지 설문이 이루어졌다. 

1기 패널이 고등학교에 진학한 7차조사부터는 전문교과(전공)에 대한 성적평가 설문이 추가되었으나 본 연구에서는 이를 분석에 포함시키지 않았다.

표 1은 각 학생들의 1차부터 9차까지의 학업성적 데이터를 통합(pooling)하여 5개 과목 성적 간 상관계수 행렬을 계산한 것이다. 예상대로 비교적 강한 양(+)의 상관성을 보이는 것을 알 수 있다. 즉, 5개 과목에 대해 (스스로의) 평가가 비슷한 경향이 있는 것이다. 이런 특징을 감안하여 우리는 이들 5개 과목 성적을 개별적으로 분석하지 않고, 이들을 평균한 종합성적(score_overall) 변수를 만들어 분석에 사용하기로 한다.

### 추정 모형

추정 모형은 다음과 같다.

$$
\text{모형 I:}\ \ \ y_{it} = \beta_0 + \beta_1 x_{it} + u_i + e_{it} \tag{1}
$$

여기서 $i$는 패널에 속한 어떤 개인을 나타내고, $t$는 개인의 특성이 관찰된 어떤 시점을 의미하며, $y_{it}$는 시점 $t$에서 개인 $i$의 학업성적을 나타낸다. $x_{it}$는 학업성적에 영향을 미치는 설명변수(벡터)로서 가구의 소득 수준이 대표적 예이다. $x_{it}$에 아래첨자 $t$가 붙어 있지만, 남녀 성별 변수와 같은 시간불변(time-invariant) 변수도 설명변수가 될 수 있다. $u_i$는 개인별 이질성(heterogeneity)을 나타내는 것으로서 확률효과(random effect) 모형에서는 이것을 확률변수로 간주한다. 

$$
\text{모형 II:}\ \ \ y_{it} = \beta_0 + \beta_1 x_{it} + \beta_2 d_t + u_i + e_{it} \tag{2}
$$

여기서 $d_t$는 연도별 특성을 포착하기 위한 시간 더미변수(벡터)로서, 가령 $d_{2011}$은 연도가 2011년이면 1의 값을 갖고, 그 외에는 0의 값을 갖는다. 

### 측정 변인 기술통계

현재 공개된 1기 다문화청소년패널조사에서 조사차수(`WAVE`)는 10차까지 있지만, 초중고 학업성적이 조사된 것은 9차까지이기 때문에 1차부터 9차까지의 데이터만 사용한다. 종속변수는 종합성적(`score_overall`)으로서 5개 과목 성적을 평균한 값이다. 

모형의 추정에 사용된 설명변수는 `WAVE`(조사차수), `male`(남학생 더미변수), `par_edu_m`(어머니 학력), `par_edu_f`(아버지 학력), `income`(월평균 소득수준), `health`(주관적 건강인식), `life_satis`(삶의 만족도), `achiev`(성취동기), `edu_plan`(희망 교육수준), `fa_support`(가족의 지지), `pr_father`(아버지에 대한 생각), `pr_mother`(어머니에 대한 생각), `parenting_a`(부모의 감독태도), `parenting_b`(부모의 방임태도), `bullying`(집단괴롭힘 피해경험), `help_in_sc`(학교 내 도움을 주는 어른의 유무 더미변수), `help_community`(학교 밖 도움을 주는 어른의 유무 더미변수) 등이다. 각 변수에 대한 자세한 설명이 표 2에 나와 있다.

## 3 패널분석 모형 I 추정 결과

우리는 패널 데이터 분석에서 일반적으로 사용되는 기본적인 모형들, 즉 통합 OLS(pooled OLS), 고정효과(FE: fixed effect), 확률효과(random effect) 등 3개 모형을 이용해 다문화청소년들의 학업성적에 영향을 미치는 요인들을 분석했다. 표 4에 추정 결과가 나와 있다. 이는 앞의 "1 연구 방법"에서 소개한 모형 I, 즉 식 1의 추정 결과이다. 

표에서 (통합 OLS 모형을 빼고) 고정효과와 확률효과 모형 위주로 추정 결과를 보면, 설정된 모형은 달라도 추정계수의 부호 및 통계적 유의성은 모형 간에 상당히 유사한 것을 알 수 있다. 특히 관심을 끄는 부분은 `income`(월평균 소득)과 `bullying`(집단괴롭힘 피해경험) 변수이다. 고정효과 및 확률효과 모형 모두 이들 두 변수의 추정계수가 1% 수준에서 유의한데, 계수의 부호가 예상과는 반대이다.  

즉, `income`의 경우, 계수가 음수로서 소득이 높은 가구의 다문화학생일수록 학업성적이 낮은 편이라는 것을 의미하며, `bullying`의 경우, 계수가 양수로서 집단괴롭힘 피해경험이 많은 학생일수록 학업성적이 우수한 경향이 있다는 것이다.  

뿐만 아니라 `help_in_sc`와 `help_community`의 경우에는 도움을 주는 어른이 각각 학교 ‘내’와 ‘밖’에 있는지를 가리키는 더미변수인데, 둘 다 도움을 주는 어른의 존재를 묻는 것으로서 차이점은 학교 내와 밖이라는 점만 다른데도 추정계수가 전자의 경우에는 음수인 반면, 후자는 양수이며, 더구나 둘 다 통계적으로 유의하다.

이런 문제와 관련하여, 우리는 III장에서 제시한 “성장” 모형이 이와 같은 문제를 해결하는 한 가지 방법이 될 수 있다고 본다. 즉, 식 (3)처럼 학업성적에 대한 설명변수로서 시간변수를 집어넣어 그것으로 학업성적의 시간에 따른 궤적을 포착(통제)해야만 `income` 등 다른 설명변수와 학업성적의 관계를 제대로 포착할 수 있다는 것이다.

## 4 패널분석 모형 II 추정 결과

패벌 회귀분석 모형에 시간 더미변수를 추가하여 추정한 결과가 표 5에 나와 있다. 이는 앞의 "1 연구 방법"에서 소개한 모형 II, 즉 식 2의 추정 결과이다. 

표에서 (통합 OLS 모형을 빼고) 고정효과와 확률효과 모형 위주로 추정 결과를 보면, 가장 먼저 시간 더미변수`wave`의 추정 계수가 `wave2`를 제외하고는 모두 1% 수준에서 유의한 것으로 나온 것을 알 수 있다. 이는 모형 I에 비해 모형 II가 더 적절하다는 것을 의미한다고 할 수 있다. 

확률효과 모형 위주로 표의 추정 결과를 차례대로 살펴보면, 먼저 남학생 더미변수(`male`)는 추정 계수(0.07)가 양수로서 통계적으로 유의하다. 다른 조건이 동일할 때, 여학생에 비해 남학생의 학업성적이 평균적으로 0.07만큼 높다는 것을 의미한다. 부모 학력이 다문화청소년의 학업성적에 미치는 영향은 어머니(`par_edu_m`)와 아버지(`par_edu_f`) 모두 정(+)의 효과로서 부모가 고학력일수록 자녀의 학업성적이 높은 경향이 있다는 것을 의미한다. 부모 중에서는 어머니에 비해 아버지 학력의 영향력이 더 큰 것으로 나타났다(추정계수 0.04 대 0.07). 

가계 소득(`income`)은 앞의 모형 I과 달리 다문화청소년의 학업성적에 유의한 영향을 미치지 않는 것으로 나타났다. 반면, 삶의 만족도(`life_satis`)와 성취동기(`achiev`)는 추정계수가 각각 0.10과 0.39로서 정(+)의 효과를 미치며, 통계적 유의성이 매우 강하고 효과의 크기도 매우 큰 편이다. 특히 성취동기의 경우 유사한 척도를 지닌 다른 변수들에 비해 학업성적에 미치는 효과가 압도적으로 큰데, 성취동기가 (4점 척도에서) 1만큼 커지면, 학업성적이 평균적으로 0.39만큼 높아지는 것으로 나타났다. 희망 교육수준(`edu_plan`) 역시 학업성적과 유의한 정(+)의 관계를 갖고 있는 것을 알 수 있다.

## 5 요약 및 시사점

다문화청소년의 초종고 학업성적이 어떤 의미를 지니고, 어떤 요인들이 어떤 식으로 학업성적에 영향을 미치는지 실증분석을 통해 살펴봤다. 주요 분석 내용을 요약하면, 첫째, 다문화청소년들의 초중고 학업성적이 나중에 대학진학과 성공적인 대학생활에 중요한 영향을 미친다. 둘째, 9년간 학업성적의 변화추이를 보면, 뚜렷한 하락세를 보인다. 셋째, 시간 효과를 통제한 모형 추정 결과, 성취동기, 삶의 만족도, 부모 학력, 가족의 지지, 부모의 관리감독, 학교 밖 도움을 주는 어른의 존재 등이 학업성적을 높이는 요인으로 작용하는 반면, 예상과 달리 경제적 요인이나 집단괴롭힘 경험 등은 유의한 영향을 미치지 않는 것으로 나타났다. 

본 연구에서 다문화청소년의 학업성적과 관련하여 여러 모형을 사용해 추정했는데, 어떤 모형을 사용했느냐에 상관없이 일관되게 통계적으로 매우 유의하고, 효과의 크기 면에서도 가장 강력한 변수를 하나만 꼽는다면 “성취동기”(`achiev` 변수)라 할 수 있다. 본 연구의 데이터에서 성취동기란 학습목표를 이루기 위해 노력하는 것, 좋아하는 일을 잘하기 위해 노력하는 것, 가치있다고 생각하는 일에 대해 최선을 다하는 것, 스스로 학습목표를 결정하는 것, 스스로 정한 목표를 향해 나아가는 것, 관심있는 일을 시도하고 즐기는 것, 자신이 만족할 때까지 노력하는 것 등을 의미하는데, 이런 성취동기가 강할수록 학업성적이 눈에 띄게 높은 경향을 보인다. 다문화청소년들의 학업성적과 관련된 사회적․정책적 대책 마련에서 이런 부분에 초점을 맞출 필요가 있다는 것을 시사한다.
