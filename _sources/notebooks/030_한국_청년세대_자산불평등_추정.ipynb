# 3 한국 청년세대 자산불평등 추정

**<a href="https://www.insnews.co.kr/m/news_view.php?firstsec=1&secondsec=17&num=69737&ckattempt=2">2030세대 자산 격차 갈수록 벌어진다</a>.**(한국보험신문, 2022-05-22)

>2030세대의 빈부격차가 갈수록 벌어지고 있다. 지난해 기준 상위 20%의 평균 자산은 10억원에 육박한 반면 하위 20%는 3,000만원에도 못 미치는 것으로 나타났다. 부의 대물림 때문이라는 분석이다.
>
>김회재 더불어민주당 의원은 통계청 가계금융복지조사 마이크로데이터를 분석한 결과를 지난 16일 발표했다. 이에 따르면 2021년 3월 말 기준 20~30대가 세대주인 가구의 평균 자산은 3억 5,651만원으로 집계됐다. 이는 전년 3억1,849만원에 비해 3802만원 늘어난 것이다.
>
>자산 규모별로 살펴보더라도 2030세대의 평균 자산은 전반적으로 늘고 있는 것으로 나타났다. 지난해 자산 규모 상위 20%의 평균 자산은 9억 8,185만원으로 2020년보다 12.8%(1억 1,141만원) 증가했다. 하위 20% 역시 2,473만원에서 2,784만원으로 12.6%(311만원) 늘었다.
>
>이처럼 2030세대 전체 평균 자산은 증가했지만 자산 격차는 심화됐다. 코로나19 이후 유동성이 늘어나면서 부동산·주식 등 자산시장이 확대된 영향으로 2030세대의 자산 5분위 배율(35.27배)이 2019년(33.21배)보다 더 커졌다. 2019년에는 상위 20% 가구의 자산이 8억원, 하위 20% 가구의 자산은 2,409만원이었다.
>
>2030세대 가구의 경상소득 5분위 배율은 2020년 기준 6.52배로 전년(6.58배)보다는 소폭 개선됐다. 소득 상위 20%의 경상소득(1억 2,832만원)이 1년 전보다 742만원(6.1%) 늘었고, 하위 20%의 경상소득(1,968만원)은 131만원(7.2%) 증가했다.
>
>작년 기준으로 자산이 상위 20%인 2030세대 가구의 전년도 평균 경상소득은 1억 592만원으로 자산 하위 20% 가구의 경상소득(3087만원)의 3.43배였다. 일각에서는 경상소득 격차 외에도 부모 등으로부터 받은 자산에 따라 청년세대 내 자산 격차가 벌려졌을 수도 있다는 지적이 나온다. 부모 등으로부터 받은 자산에서 근본적인 차이가 날 수 있다는 의미다. 김 의원은 “소득 격차만으로는 35배에 달하는 자산 격차를 설명하기 어렵다. 삶의 출발선에서부터 극복하기 어려운 격차를 안고 시작하는 것은 공정하지 못하다”면서 “경제적 불평등을 해소하는 정의로운 전환이 절실하다. 부모 찬스가 없는 청년들도 격차를 해소할 수 있도록 공정의 사다리가 마련돼야 한다”고 강조했다.
>
- (청년기본법상 청년의 법정 연령 기준은 만 19~34세다.)
