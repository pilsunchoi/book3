### 내용

* 캐글러(Kaggler)를 대상으로 한 설문조사(Kaggle은 데이터 사이언스 커뮤니티임)

* 특정 국가 또는 지역에서 응답자가 50명 미만인 경우 익명을 위해 그룹을 '기타'로 처리
* 취업 상태에 관한 질문에 답변하지 않은 응답자는 제외(이 질문은 첫 번째 필수 질문이기에 응답하지 않으면 응답자가 다섯 번째 질문 이후 진행되지 않음)
* 모든 질문이 모든 응답자에게 보여지는 것은 아님. 취업을 한 사람과 학생을 나누어 다른 질문을 함
* 응답자의 신원을 보호하기 위해 주관식과 객관식 파일로 분리
* 객관식과 자유 형식 응답을 맞추기 위한 키를 제공하지 않음
* 주관식 응답은 같은 행에 나타나는 응답이 반드시 동일한 설문조사자가 제공하지 않도록 열 단위로 무작위 지정

#### 참고: IRB(Institutional Review Board, 임상시험 심사위원회)

출처: [실험대 위에 놓인 연구윤리: 임상시험](https://post.naver.com/viewer/postView.naver?volumeNo=32142480&memberNo=40871286&vType=VERTICAL)
- IRB는 인간을 대상으로 하는 모든 의학·과학 연구의 윤리적·과학적 측면을 심사하는 기관입니다. 의학, 약학, 법학, 통계학 등 다양한 분야의 전문가로 구성되어 임상 시험이 필요한 병원이나 연구 기관에 설치되지요. IRB는 실험의 목적이 충분히 가치 있는지, 피실험자에게 제공되는 동의서의 내용은 적절한지, 실험 방법이 윤리적으로 타당한지 등을 면밀히 검토한 후 임상 시험을 승인합니다.


- 현재 우리나라에서 ‘살아 있는 사람’이 포함되는 모든 연구는 IRB의 승인을 받아야 합니다. 여기에는 사람의 몸에 직접 하는 실험뿐 아니라 소변 검사나 신체에 대한 설문 조사 등도 해당합니다. 고작 소변을 분석하거나 설문지 답변을 참고하는 것이 왜 문제가 되는지 의문이 들 수도 있어요. 하지만 인간에게서 얻은 모든 표본, 예를 들어 소변 한 컵에도 그 사람의 DNA를 비롯해 수많은 신체 정보가 들어 있습니다. 설문 조사도 마찬가지로 다양한 개인 정보를 포함하지요. 이를 타당한 근거와 당사자의 허락 없이 사용하는 것은 엄연한 인권 침해입니다.

* [RespondentTypeREADME.txt](https://raw.githubusercontent.com/corazzon/KaggleStruggle/master/kaggle-survey-2017/data/RespondentTypeREADME.txt): schema.csv 파일의 "Asked" 열 응답을 디코딩하는 스키마이다.

```python
# This Python 3 environment

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
import glob
import warnings
import squarify
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.patches as mpatches
from matplotlib import ticker
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Load Dataset
responses_data = pd.read_csv("../Data/kaggle_survey_2021_responses.csv", 
                             dtype=object, low_memory=False)
```

## 응답자 인구통계

### 국가별 응답자 수
Q: 귀하는 현재 어느 나라에 거주하고 있습니까?

```python
plt.rcParams['figure.dpi'] = 300
fig = plt.figure(figsize=(3, 9), facecolor='#FEFBF3')
gs = fig.add_gridspec(2, 1)
gs.update(wspace=1.5, hspace=1.1)

##########COUNTRY##########
temp = pd.DataFrame(responses_data.groupby('Q3', dropna=False).size()).reset_index()
temp.columns = ['country', 'amount']
temp['pct'] = temp['amount']*100/temp.amount.sum()
temp = temp.sort_values('amount', ascending=False).reset_index()
temp = temp.query("amount>1")
temp.loc[1, 'country'] = 'USA'
temp.loc[8, 'country'] = 'UK'
temp.loc[27, 'country'] = 'Iran'
for i in range(31,66):
    temp.at[i,'country'] = 'Other'

temp = temp.groupby('country').sum()
temp = temp.query("country!='Other'")
temp = temp.query("pct>1")
temp = temp.sort_values('amount', ascending=False).reset_index()
temp = temp[0:20]

background_color = "#FEFBF3"
color_map = ["#90AACB" for _ in range(20)]
color_map[0] = "#F2A154" #F2A154 79B4B7
sns.set_palette(sns.color_palette(color_map))

ax0 = fig.add_subplot(gs[0, 0])
ax0.set_facecolor(background_color)
ax0.set(xlim=(0, 40))
ax0.set(ylim=(0, 20))
for s in ["right", "top"]:
    ax0.spines[s].set_visible(False)

#graph
ax0_sns = sns.barplot(ax=ax0, y=temp['country'], x=temp['pct'], 
                      zorder=2, linewidth=0, orient='h', saturation=0.9, alpha=0.7)
ax0_sns.grid(which='major', axis='x', zorder=0, color='#EEEEEE', linewidth=0.2)

#format axis
ax0_sns.set_xlabel("Percentage",fontsize=4, weight='semibold')
ax0_sns.set_ylabel("Country",fontsize=4, weight='semibold')
ax0_sns.tick_params(labelsize=4, width=0.3, length=1)

#title
x0, x1 = ax0.get_xlim()
y0, y1 = ax0.get_ylim()
ax0.text(x0, y1-1.5, 'Top 20 Countries', fontsize=5, ha='left', va='top', weight='bold')
ax0.text(x0, y1-0.8, 'Q3: In which country do you currently reside?', 
         fontsize=4, ha='left', va='top')

# data label
i=0
for p in ax0.patches:
    value = f'{p.get_width():.1f}%'
    if i<7:
        x1 = -1.3
    else:
        x1 = 1.0
    x = p.get_x() + p.get_width() + x1
    i+=1
    y = p.get_y() + p.get_height() / 2 
    ax0.text(x, y, value, ha='center', va='center', fontsize=3, weight='semibold') 

x_format = ticker.PercentFormatter()
ax0.xaxis.set_major_formatter(x_format)

plt.show()
```

```python
##########COUNTRY##########
temp = pd.DataFrame(responses_data['Q3'])
temp = temp.drop(0)
temp.columns = ['country']
countries = temp['country'].apply(lambda x: 0 if pd.isnull(x) else x)
text = " ".join(country for country in countries.astype(str))

from wordcloud import WordCloud, STOPWORDS
stopwords = set(STOPWORDS)

x, y = np.ogrid[:400, :400]
mask = (x - 200) ** 2 + (y - 200) ** 2 > 180 ** 2
mask = 255 * mask.astype(int)

word_cloud = WordCloud(stopwords = stopwords, max_words=100, collocations = True, 
                       repeat = True, background_color = '#FEFBF3', 
                       width = 1920, height = 640, mask=mask)
word_cloud.generate(text)

# saving the image
word_cloud.to_file('got.png')

# Display the generated Word Cloud
plt.rcParams['figure.dpi'] = 900
plt.figure(figsize=(1, 1))
plt.tight_layout(pad=0)
plt.imshow(word_cloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

응답자 중 인도 거주자의 비중이 28.6%로 최대 다수를 차지하고 있다. 2위는 미국(10.2%)으로 인도와 거의 20%포인트의 큰 격차를 보이고 있으며, 3위와 또 다른 큰 격차는 일본, 중국, 브라질, 러시아, 나이지리아의 사용자가 약 3%로 거의 같은 비율로 채웠습니다. 유럽 사용자는 러시아, 영국, 독일, 스페인, 프랑스 및 이탈리아(포함)이며 각각 2.8%, 2.1%, 1.8%, 1.7%, 1.5% 및 1.2%입니다. 동남아시아의 사용자는 상위 20개 국가 목록에서 인도네시아(1.7%)뿐입니다.

```python
plt.rcParams['figure.dpi'] = 300
fig = plt.figure(figsize=(5, 4), facecolor='#FEFBF3')
gs = fig.add_gridspec(2, 2)
gs.update(wspace=0.2, hspace=1.5)

##########SEXPIE##########
temp = pd.DataFrame(responses_data.groupby('Q2', dropna=False).size()).reset_index()
temp.columns = ['answer', 'amount']
temp.loc[(temp.answer=='Nonbinary') | (temp.answer=='Prefer not to say')
         | (temp.answer=='Prefer to self-describe'), 'answer'] \
        = 'Prefer not to say/self-describe'
temp = temp.groupby('answer').sum().reset_index()
temp = temp.drop(2)
temp = temp.sort_values('amount', ascending=False)
color_map = ['#F2A154', '#90AACB', 'gray', 'darkgray', 'lightgray']
ax0 = fig.add_subplot(gs[0, 0])
ax0.set_facecolor(background_color)

#graph
ax0.pie(x=temp['amount'], wedgeprops=dict(width=0.12), 
        colors=color_map, pctdistance=0.7,
        textprops={'fontsize': 3, 'weight': 'semibold'}, autopct='%1.1f%%')

#title
x0, x1 = ax0.get_xlim()
y0, y1 = ax0.get_ylim()
ax0.text(x0, y1*1.60, 'Respondents by Gender', fontsize=5, ha='left', va='top', 
         weight='bold')
ax0.text(x0, y1*1.45, 'What is your gender?', fontsize=3, ha='left', va='top')
ax0.legend(temp['answer'], loc="upper left", bbox_to_anchor=(x0*0.03, y1*0.92), 
           prop={'size': 3}, frameon=False, ncol=3)

#format tick
ax0.tick_params(labelsize=4, width=0.5, length=1.5)

##########SEXBAR##########
temp = pd.DataFrame(responses_data.groupby('Q2', dropna=False).size()).reset_index()
temp.columns = ['answer', 'amount']
temp['pct'] = temp['amount']*100/temp.amount.sum()
temp = temp.drop(4)
temp = temp.sort_values('pct', ascending=False)

background_color = "#FEFBF3"
color_map = ["#90AACB" for _ in range(75)]
color_map[0] = "#F2A154"
sns.set_palette(sns.color_palette(color_map))

ax0 = fig.add_subplot(gs[0, 1])
ax0.set_facecolor(background_color)
ax0.set(xlim=(0, 5))
ax0.set(ylim=(0, 100))
for s in ["right", "top"]:
    ax0.spines[s].set_visible(False)

#graph
ax0 = sns.barplot(ax=ax0, y=temp['pct'], x=temp['answer'], zorder=2, 
                  linewidth=0, saturation=.9, alpha=0.7)
fillwidth = np.array([-0.5,1.5])
ax0.fill_between(fillwidth, np.max(temp['pct']+2), color='#FFE699', 
                 alpha=0.5, zorder=2, linewidth=0)

for s in ["right", "top"]:
    ax0.spines[s].set_visible(False)

ax0.grid(which='major', axis='y', zorder=0, color='#EEEEEE', lw=0.3)

#title
x0, x1 = ax0.get_xlim()
y0, y1 = ax0.get_ylim() 

# data label
i=0
for p in ax0.patches:
    value = f'{p.get_height():.1f}%'
    if i<2:
        y1 = -3
    else:
        y1 = 3
    i+=1
    x = p.get_x() + p.get_width() / 2 
    y = p.get_y() + p.get_height() + y1
    ax0.text(x, y, value, ha='center', va='center', fontsize=3, weight='semibold') 

#format axis
ax0.set_xlabel("Gender",fontsize=4, weight='semibold')
ax0.set_ylabel("Percentage",fontsize=4, weight='semibold')
ax0.tick_params(labelsize=3.0, width=0.3, length=1)
#plt.xticks(rotation=90)

y_format = ticker.PercentFormatter()
ax0.yaxis.set_major_formatter(y_format)

plt.show()
```

### 프로그래밍 언어
어떤 프로그래밍 언어를 정기적으로 사용합니까?


파이썬(Python)이 데이터 과학자를 포함한 모든 개발자에게 사랑받는 프로그래밍 언어의 스위스 군대임을 과감하게 보여줍니다. Python은 배우기 쉽고 거의 모든 것을 할 수 있습니다. 두 번째 언어는 SQL(16.3%)입니다. SQL은 데이터와 대화하는 것입니다. 데이터 과학 세계에서 SQL은 필수 지식입니다. 예, 이 노트북은 물론 Python으로 작성되었습니다.

```python
##########PROGRAMMINGLANGUAGE##########
col = 7
result=pd.DataFrame()
for col in range(col, col+13):
    choice = responses_data.iloc[0:1, col:col+1].to_string()
    lang = choice[choice.find('ce -')+5:len(choice)]
    grouped = responses_data.iloc[:, -1].groupby(responses_data.iloc[:, col])
    data = {"language": lang, "amount": grouped.size()[0]}
    result = result.append(data, ignore_index=True)

total = result['amount'].sum()
result = result.query("language!='None' & language!='Other'")
result['pct'] = result['amount']*100/total
temp = result
temp = temp.sort_values('pct', ascending=False)

plt.rcParams['figure.dpi'] = 300
fig = plt.figure(figsize=(3, 8), facecolor='#FEFBF3')
gs = fig.add_gridspec(2, 1)
gs.update(wspace=1.5, hspace=1.1)

background_color = "#FEFBF3"
color_map = ["#90AACB" for _ in range(75)]
color_map[0] = "#F2A154"
sns.set_palette(sns.color_palette(color_map))

ax0 = fig.add_subplot(gs[1, 0])
ax0.set_facecolor(background_color)
ax0.set(xlim=(0, 40))
ax0.set(ylim=(0, 11))
for s in ["right", "top"]:
    ax0.spines[s].set_visible(False)

#graph
ax0 = sns.barplot(ax=ax0, y=temp['language'], x=temp['pct'], 
                      zorder=2, linewidth=0, orient='h', saturation=0.9, alpha=0.7)

ax0.grid(which='major', axis='x', zorder=0, color='#CDD0CB', linewidth=0.3, alpha=0.5)
ax0.grid(which='major', axis='y', zorder=0, color='#E8EAE6', linewidth=10.5, alpha=0.7)

#format axis
ax0.set_xlabel("Percentage",fontsize=4, weight='semibold')
ax0.set_ylabel("Programming Language",fontsize=4, weight='semibold')
ax0.tick_params(labelsize=4, width=0.3, length=1)

#title
x0, x1 = ax0.get_xlim()
y0, y1 = ax0.get_ylim()
ax0.text(x0, y1-1.5, 'Programming Language', fontsize=5, ha='left', 
         va='top', weight='bold')
ax0.text(x0, y1-1, 'What programming languages do you use on a regular basis?', 
         fontsize=4, ha='left', va='top')

# data label
i=0
for p in ax0.patches:
    value = f'{p.get_width():.1f}%'
    if i<7:
        x1 = -1.4
    else:
        x1 = 1.2
    
    i+=1
    x = p.get_x() + p.get_width() + x1
    y = p.get_y() + p.get_height() / 2 
    ax0.text(x, y, value, ha='center', va='center', fontsize=3.5, weight='semibold') 

x_format = ticker.PercentFormatter()
ax0.xaxis.set_major_formatter(x_format)

plt.show()
```
