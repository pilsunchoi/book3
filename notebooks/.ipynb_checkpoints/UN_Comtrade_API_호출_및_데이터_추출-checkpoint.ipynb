## UN Comtrade 데이터베이스

### 소개

UN Comtrade 데이터베이스는 세계 각국의 무역 데이터를 수집하고 제공하는 주요 소스 중 하나로, 다양한 특징과 장점이 있다.

**특징**

1. 광범위한 데이터 커버리지:
    - 전 세계 대부분의 국가와 지역의 무역 데이터를 포함하며, 연도별, 월별 데이터를 제공한다.
    - 상품 분류 체계로 HS (Harmonized System), SITC (Standard International Trade Classification), BEC (Broad Economic Categories) 등 다양한 상품 분류 기준으로 데이터를 제공한다.

> **HS**(Harmonized System)는 국제적으로 통일된 상품 분류 체계이다. 세계관세기구(WCO)가 개발했으며, 상품에 6자리 코드를 부여하여 전 세계적으로 무역 통계를 수집하고 분석하는 데 사용된다. SITC(Standard International Trade Classification)는 유엔통계국(UN Statistics Division)이 개발한 분류 체계로, 국제 무역 데이터를 분석하고 비교하는 데 사용된다. SITC는 상품을 경제적 활동에 맞게 그룹화하여 1자리에서 5자리 코드로 세분화한다. BEC(Broad Economic Categories) 역시 유엔통계국이 개발한 분류 체계로, 상품과 서비스를 경제적 용도에 따라 분류한다. BEC는 주로 자본재, 중간재, 소비재로 나누며, 경제 분석과 무역 연구에 활용된다.

2. 상세한 데이터:
    - 무역 데이터는 국가 간의 수출입 활동을 다양한 수준(총계, 세부 품목 등)에서 보여준다.
    - 수출입 국가, 무역 품목, 무역 흐름(수출, 수입, 재수출, 재수입) 등을 세부적으로 구분하여 제공한다.

3. 정기적인 업데이트:
    - 데이터는 정기적으로 업데이트되며, 최신 무역 동향을 파악할 수 있다.
    - 각 국가의 무역 보고서가 제공될 때마다 데이터가 갱신된다.

**정확성**

1. 데이터 수집 및 검증:
    - UN Comtrade는 각국의 관세청 및 통계청으로부터 데이터를 수집한다.
    - 수집된 데이터는 검증 과정을 거치며, 국제 기준에 따라 표준화된다.


2. 데이터 품질 관리:
    - 데이터 품질을 유지하기 위해 여러 단계의 검증과 교차 확인 절차를 거친다.
    - 오차나 불일치가 있는 경우, 관련 기관과 협력하여 데이터를 수정 및 보완한다.

**공공성**

1. 무료 접근 가능성:
    - UN Comtrade 데이터는 대부분 무료로 접근할 수 있으며, 누구나 사용할 수 있다.
    - 유료인 프리미엄 서비스는 추가적인 기능과 더 빠른 데이터 접근을 제공한다.

2. 투명성:
    - 데이터 수집 및 처리 과정이 투명하며, 사용자는 데이터의 출처와 처리 과정을 확인할 수 있다.

**유용성**

1. 연구 및 분석 도구:
    - 연구자, 정책입안자 등이 무역 흐름을 분석하고 경제 정책을 수립하는 데 유용하다.
    - 무역 패턴, 비교 우위, 시장 동향 등을 분석하는 데 필수적인 데이터 소스이다.

2. 비즈니스 인텔리전스:
    - 기업은 UN Comtrade 데이터를 활용하여 시장 진출 전략을 세우고, 경쟁 환경을 분석하며, 무역 기회를 발굴할 수 있다.

3. 정책 결정 지원:
    - 정부와 국제기구는 이 데이터를 사용하여 무역 정책을 수립하고, 경제 협력을 증진하며, 무역 장벽을 이해하고 대응할 수 있다.

**참고**

UN Comtrade에 대한 자세한 내용은 [UN Comtrade Docs](https://comtrade.un.org)에서 확인할 수 있다.

### API를 통해 UN Comtrade 데이터를 효율적으로 다운로드 받기

UN Comtrade 데이터는 [웹사이트](https://comtradeplus.un.org/)를 통해 직접 접근할 수 있으며, 다양한 형식(CSV, Excel 등)으로 다운로드받을 수 있다.(드롭다운 메뉴를 통해 원하는 대상 및 항목을 선택하는 방식이다.) 

그러나 UN Comtrade 데이터를 보다 효율적으로 다운로드 받는 방법은 API(Application Programming Interface)를 통해 프로그래밍적으로 데이터를 가져오는 것이다. API를 통해 데이터를 다운로드 받는 것은 API를 사용해 상대방 서버에 데이터를 프로그래밍적으로 요청하고, 그 응답을 받아오는 과정을 의미한다. API는 소프트웨어 간의 상호 작용을 가능하게 하는 인터페이스로, 이를 사용함으로써 수동으로 데이터를 다운로드할 필요 없이, 코드로 데이터를 요청하고 처리할 수 있어 효율적이다.

**UN Comtrade API 구독키 획득**

API Key는 API를 사용할 때 사용자를 인증(확인)하고 해당 사용자가 어떤 데이터나 서비스를 사용할 수 있는지 권한을 부여하기 위한 고유한 식별자이다. 이는 마치 웹사이트에 로그인할 때 사용하는 사용자 이름과 비밀번호와 같은 역할을 한다. 

   - UN Comtrade API를 사용하려면 구독키가 필요하다.
   - 구독키를 얻기 위해서는 [UN Comtrade API portal](https://comtradedeveloper.un.org)에 방문한다. 아래 그림에서 왼쪽 중간의 로그인 단추를 클릭하면 새로운 창이 생기면서 로그인 정보의 입력을 요구한다. 
   
   
   
   - 이미 @un.org 계정이 있거나 여기에 새로이 가입하는 경우 이를 통해 로그인할 수 있다.
   - 또는 자신의 Gmail 계정과 비밀번호를 사용하여 UN Comtrade에 계정을 설정할 수 있다.
   - 로그인 후 아래와 같이 상단 메뉴에서 `Products`를 선택한다.
   
   
   
   - 프리미엄 개인 또는 기관 등급이 아닐 경우 `Free APIs`를 선택한다. 
   - 원하는 구독 이름을 만들고 `Subscribe`를 클릭하면 구독키가 생성된다.
   - 구독키가 발급되면 상단 메뉴의 `Profile`에서 확인할 수 있다.
   - 구독키가 사용 도중 중단되는 경우, `Profile`에서 재발급받을 수 있다.

## UN Comtrade API 호출 파이썬 패키지

### comtradeapicall 개요
[comtradeapicall](https://pypi.org/project/comtradeapicall/)는 파이썬 패키지로서 [UN Comtrade API](https://comtradedeveloper.un.org/signin?returnUrl=%2F)를 호출하여 데이터를 추출하고 다운로드하는 작업을 쉽게 수행할 수 있도록 해준다. 

**[데이터베이스별 API 종류](https://comtradedeveloper.un.org/apis)

- UN Comtrade 데이터베이스별로 무료 또는 유료(프리미엄) 구독키가 필요하고, 구독키 없이 접근 가능한 경우도 있다.

| 데이터베이스         | 내용                     |
|:-------------------|:-----------------------------------------|
| comtrade - v1      | Free API. 무역 데이터 및 메타데이터. 구독키가 필요함. |
| comtradeAsync - v1 | Premium API. 비동기 요청(asynchronous request). 이메일로 결과 알림을  받음. 구독키가 필요함. |
| comtradebulk - v1  | Premium API. 대용량 파일. 구독키가 필요함. |
| experimental - v1  | AIS(선박자동식별시스템) 데이터를 사용하여 해상 무역을 추정한 실험적 데이터.   |
| public - v1        | Free API. 무역 데이터 미리보기. 구독키가 필요하지 않음.    |

### 함수 파라미터

**SelectionCriteria**
- `typeCode`(str) : 상품 유형. 상품 (C) 또는 서비스 (S)
- `freqCode`(str) : 관측이 발생하는 시간 간격. 연간 (A) 또는 월간 (M)
- `clCode`(str) : 사용되는 상품 분류와 그 버전을 나타냄 (HS, SITC)
- `period`(str) : 월간의 경우 연도와 월의 조합, 연간의 경우 연도
- `reporterCode`(str) : 측정된 통계 현상이 관련된 국가 또는 지리적 영역
- `cmdCode`(str) : 분류 코드와 함께 사용하는 상품 코드
- `flowCode`(str) : 무역 흐름 또는 하위 흐름 (수출, 재수출, 수입, 재수입 등)
- `partnerCode`(str) : 해당 무역 흐름의 주요 상대국 또는 지리적 영역
- `partner2Code`(str) : 해당 무역 흐름의 보조 상대국 또는 지리적 영역
- `customsCode`(str) : 세관 또는 통계 절차
- `motCode`(str) : 상품이 국가의 경제 영토를 드나들 때 사용되는 운송 방식

**query_option**
- `maxRecords`(int) : 반환되는 레코드 수 제한
- `format_output`(str) : 출력 형식. CSV 또는 JSON
- `aggregateBy`(str) : 쿼리 집계 옵션
- `breakdownMode`(str) : classic 모드(상대국/상품별 무역) 또는 plus 모드(확장된 세분화) 선택 옵션
- `countOnly`(bool) : `True`로 설정하면 실제 레코드 수 반환
- `includeDesc`(bool) : 설명 포함 여부 옵션

**AIS SelectionCriteria**
- `typeCode`(str) : 상품 유형. 상품 (C)만 해당
- `freqCode`(str) : 시간 간격. 일간 (D)

## UN Comtrade 데이터 분석 예제

### 집중성 마진과 확장성 마진 계산
    
Hummels와 Klenow(2005)는 무역 데이터 분석에서 두 가지 주요 개념인 집중성 마진(intensive margin)과 확장성 마진(extensive margin)을 도입했다. 이 개념들은 나라들의 무역 성과를 더 정교하게 평가하기 위해 사용된다.(Hummels, D., & Klenow, P. J. (2005). The variety and quality of a nation's exports. *American economic review*, 95(3), 704-723.)

>참고: 본 주제에 대한 Stata 코드는 UNCTAD에서 발간한 [A Practical Guide to Trade Policy Analysis](https://unctad.org/topic/trade-analysis/trade-policy-analysis)의 [CHAPTER 1: Analyzing trade flows](https://unctad.org/system/files/official-document/gds2012d2_ch1_en.pdf) pp.24-26, "Box 1.1 Intensive and extensive margins of diversification"을 참조할 것.

**집중성 마진**
어떤 나라의 집중성 마진은 그 나라가 이미 수출하고 있는 제품들에서 시장 점유율이 어느 정도인지를 측정한 것이다. 어떤 나라 $i$의 집중성 마진을 식으로 표현하면 다음과 같다.

$$
IM_i = \frac{\sum_{k \in K_i} X_{i,k}}{\sum_{k \in K_i} X_{W,k}} \tag {4.1}
$$

여기서, $K_i$는 $i$국이 수출하는 제품의 집합이고, $X_{i,k}$는 $i$국의 $k$재 수출액이며, $X_{W,k}$는 세계 전체의 $k$재 수출액이다. 따라서 위 식의 분자는 $i$국의 총 수출액이고, 분모는 $i$국 수출 상품에 대해 세계 각국 수출액을 모두 합친 것이다. 

**확장성 마진**
확장성 마진은 어떤 나라가 수출하는 제품들이 세계 무역에서 차지하는 비중을 측정한 것이다. 이는 해당 국가가 새로운 제품을 수출하거나 새로운 시장에 진출함으로써 무역을 확장할 수 있는 잠재력을 나타낸다. 식으로 표현하면 다음과 같다.

$$
EM_i = \frac{\sum_{k \in K_i} X_{W,k}}{\sum_{k \in K_W} X_{W,k}}  \tag {4.2}
$$

여기서, $K_W$는 모든 수출 상품의 집합이다. 따라서 위 식의 분자는 $i$국 수출 상품에 대해 세계 각국 수출액을 모두 합친 것이고, 분모는 모든 수출 상품에 대해 세계 각국 수출액을 모두 합친 것이다. 

**의의**

Hummels와 Klenow의 집중성 마진과 확장성 마진은 전통적인 무역 측정 방법에 비해 몇 가지 중요한 이점을 제공한다. 

- 기존의 단순한 무역량 분석에서 벗어나, 특정 국가가 기존 수출품목을 통해 얼마나 많은 시장 점유율을 차지하고 있는지를 파악할 수 있다. 이는 특정 산업이나 제품군에서의 경쟁력을 평가하는 데도 유용하다. 

- 국가가 새로운 제품을 수출하거나 새로운 시장에 진출하는 능력을 평가할 수 있다. 이는 정책 결정자들이 무역 다변화 전략을 수립하는 데 중요한 정보를 제공한다.

- 집중성 마진은 기존 제품의 시장 점유율 증대를 목표로 하는 정책의 효과를 측정할 수 있으며, 확장성 마진은 새로운 시장과 제품을 개척하는 정책의 효과를 평가하는 데 유용하다.

### 간단한 숫자 예

```python
# Creating the DataFrame
data = {
    'reporter': ['KOR', 'USA', 'USA'],
    'partner': ['WLD', 'WLD', 'WLD'],
    'year': [2000, 2000, 2000],
    'product': ['자동차', '자동차', '농산물'],
    'trade_value': [3, 5, 2]
}
df = pd.DataFrame(data)
df
```

**한국의 IM 및 EM**

- 위 예에서 한국(KOR)의 집중성 마진은 (한국이 수출하는) 자동차 수출시장에서 한국의 세계 시장 점유율로서 $\frac{3}{3+5}=0.375$, 즉 37.5%이다.
- 한국의 확장성 마진은 세계 전체 수출 시장에서 (한국이 수출하는) 자동차 수출시장의 점유율로서 $\frac{3+5}{3+5+2}=0.8$, 즉 80%이다.

**미국 IM 및 EM**

- 위 예에서 미국(USA)의 집중성 마진은 (미국이 수출하는) 자동차 및 농산물 수출시장에서 미국의 세계 시장 점유율로서 $\frac{5+2}{3+5+2}=0.7$, 즉 70%이다.
- 미국의 확장성 마진은 세계 전체 수출시장에서 (미국이 수출하는) 자동차 및 농산물 수출시장이 차지하는 점유율로서 $\frac{3+5+2}{3+5+2}=1$, 즉 100%이다.


**다음 코딩에 의한 계산 결과와 일치하는 것을 확인할 수 있다.**

```python
df['x_i_k'] = df['trade_value'] #각국의 상품별 수출
df['temp1'] = df.groupby(['year', 'product'])['x_i_k'].transform('sum') #k재의 연도별 세계수출액
df['sum_i_x_i_k'] = df.groupby(['reporter', 'year'])['x_i_k'].transform('sum') #각국의 상품별 수출을 각국별로 합침
df['sum_i_x_w_k'] = df.groupby(['reporter', 'year'])['temp1'].transform('sum') 
df['sum_w_x_w_k'] = df.groupby('year')['x_i_k'].transform('sum') #각국의 상품별 수출을 모두 합침

df['im_i'] = df['sum_i_x_i_k'] / df['sum_i_x_w_k']
df['em_i'] = df['sum_i_x_w_k'] / df['sum_w_x_w_k']
df
```

```python
# HS 6단위 기준으로 집중성 마진과 확장성 마진을 계산하기 위해 'aggrLevel' 값이 6인 관측만 선택

df1 = df[df['aggrLevel'] == 6]
```

```python
df1 = df1[['period', 'reporterISO', 'reporterDesc', 'partnerISO', 'partnerDesc',
           'cmdCode', 'cmdDesc', 'fobvalue']]
```

```python
df1.columns = ['year', 'reporter', 'reporter_name', 'partner', 'partner_name', 
               'product', 'product_name', 'trade_value']
df1
```

```python
# Sectoral decomposition
# Construct Extensive and Intensive margin of diversification
df1['x_i_k'] = df1['trade_value'] #각국의 상품별 수출
df1['temp1'] = df1.groupby(['year', 'product'])['x_i_k'].transform('sum') #k재의 연도별 세계수출액
df1['sum_i_x_i_k'] = df1.groupby(['reporter', 'year'])['x_i_k'].transform('sum') #각국의 상품별 수출을 각국별로 합침
df1['sum_i_x_w_k'] = df1.groupby(['reporter', 'year'])['temp1'].transform('sum') 
df1['sum_w_x_w_k'] = df1.groupby('year')['x_i_k'].transform('sum') #각국의 상품별 수출을 모두 합침

df1['im_i'] = df1['sum_i_x_i_k'] / df1['sum_i_x_w_k']
df1['em_i'] = df1['sum_i_x_w_k'] / df1['sum_w_x_w_k']
df1
```

```python
# 필요 변수만 선택하고 'im_i', 'em_i'에 100을 곱함
df2 = df1[['reporter', 'im_i', 'em_i']].copy()
df2['im_i'] = df2['im_i'] * 100
df2['em_i'] = df2['em_i'] * 100

# 중복 제거
df2 = df2.drop_duplicates()

# 결과 통계량
df2.describe()
```

### AIS 기반 해상 무역량 데이터

UN Comtrade 데이터베이스에는 AIS(선박자동식별시스템)<sup>[주7](#footnote7)</sup> 데이터를 사용하여 해상 무역을 추정하는 실험적 데이터가 있다. IMF(국제통화기금)의 Cerdeiro 등의 연구진이 개발한 것이다.([Cerdeiro, D. A., Cerdeiro, M. D. A., Komaromi, A., Liu, Y., & Saeed, M. (2020). World seaborne trade in real time: A proof of concept for building AIS-based nowcasts from scratch (No. 20-57). International Monetary Fund.](https://www.imf.org/-/media/Files/Publications/WP/2020/English/wpiea2020057-print-pdf.ashx))  
    <small> <p id="footnote6"> [주7]: AIS(Automatic Identification System)은 선박이 자신의 위치, 속도, 항로 등의 정보를 주기적으로 발신하여 다른 선박 및 해안 당국과 공유하는 시스템이다. 2002년 국제해사기구(IMO: International Maritime Organization)에 의해 도입되었다. 이후 대부분의 상업용 선박과 큰 여객선에 대해 AIS 탑재가 의무화되었다. 현재 AIS는 전 세계적으로 널리 사용되고 있으며, 해양 안전을 강화하고 해상 교통 관리, 사고 대응, 환경 모니터링, 무역 분석 등 다양한 분야에서 중요한 역할을 하고 있다. </p> </small>

해상 운송은 세계 무역의 핵심이며 실시간으로 추적할 수 있다. 유엔무역개발회의(UNCTAD, 2017)에 따르면 전 세계 상품 무역의 80% 이상이 해상 운송을 통해 이루어지며, 금액으로는 70% 이상을 차지한다. 화물선에는 주기적으로 신호를 전송하는 장치(AIS)가 장착되어 있으며, 선박의 위치, 속도, 흘수(draught) 등의 정보가 들어있다.  Cerdeiro 등(2020)은 AIS 원시 데이터를 사용하여 세계 해상 무역 지표를 구축했다. 절차를 보면, 우선 AIS 메시지에 공간 클러스터링(spatial clustering) 알고리즘을 적용하여 잠재적인 입항 데이터 세트를 구축한다. 이 잠재적인 입항 세트를 공식적인 선박 수준의 미국 입항 기록 데이터와 결합시킨다. 이 공식 데이터를 사용하여 특정 잠재적 입항이 실제 입항인지 여부를 예측하는 분류기를 훈련시킨다. 이를 통해 두 항만 간의 무역량을 추정하고 수입, 수출, 국내 무역량을 식별한다. 요컨대, AIS 데이터에 머신러닝 기법을 적용하여 세계 전체, 양국, 그리고 국가 내 수준에서 무역량을 추정해내는 것이다.

### 2024년 1-4월 DRY BULK 해양 수출 상위 10개국(AIS 기반 추정)

```python
# 조건에 맞는 데이터 필터링
filtered_df = df_aistrade_aggregated[
    (df_aistrade_aggregated['year_month'] == '2024-01') &\
    (df_aistrade_aggregated['flowCode'] == 'X')]

# countryareaDesc별로 그룹화한 후 mtc 값을 합산
aggregated_df = filtered_df.groupby('countryareaDesc')['mtc'].sum().reset_index()

# mtc 전체 합계 계산
aggregated_df['total_mtc'] = aggregated_df['mtc'].sum()

# mtc_ratio 컬럼 생성
aggregated_df['mtc_ratio'] = (aggregated_df['mtc'] / aggregated_df['total_mtc']) * 100

# mtc_ratio를 기준으로 내림차순 정렬
aggregated_df = aggregated_df.sort_values(by='mtc_ratio', ascending=False)

# 결과 출력
aggregated_df.head(10)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 처음 10개 countryareaDesc를 선택하고 나머지는 'Others'로 표시
top_10 = aggregated_df.nlargest(10, 'mtc_ratio')
Others = aggregated_df[~aggregated_df['countryareaDesc'].isin(top_10['countryareaDesc'])]

# 'Others' 행 추가
Others_sum = Others['mtc_ratio'].sum()
Others_row = pd.DataFrame({'countryareaDesc': ['Others'], 'mtc_ratio': [Others_sum]})
top_10 = pd.concat([top_10, Others_row], ignore_index=True)

# 'Others'를 맨 마지막으로 이동
top_10 = top_10[top_10['countryareaDesc'] != 'Others']
top_10 = pd.concat([top_10, Others_row], ignore_index=True)

# 도넛 차트 생성
fig, ax = plt.subplots(figsize=(8, 5), dpi=500)

# 색상 설정 (마지막 'Others'를 회색으로)
colors = plt.cm.tab20.colors[:len(top_10)-1] + ('lightgray',)

# 파이 차트 생성 및 도넛 모양으로 만들기 위해 중심에 구멍을 만듦
wedges, texts, autotexts = ax.pie(top_10['mtc_ratio'], labels=None, autopct='%1.0f%%',
                                  startangle=90, counterclock=False, colors=colors, wedgeprops=dict(width=0.3), pctdistance=0.85)

# 파이 조각의 값을 하얀색으로 표시
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontsize(10)
    autotext.set_ha('center')
    autotext.set_va('center')

# 제목과 레전드 설정
plt.title('Top 10 Countries for DRY BULK Seaborne Exports (Jan-Apr 2024)')
plt.legend(wedges, top_10['countryareaDesc'], title="Country", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

# 레이아웃 조정
plt.tight_layout()

# 도표 표시
plt.show()
```
