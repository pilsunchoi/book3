## 1 서론

소득이나 자산의 분포 및 불평등을 측정하는 것은 실질적, 정책적 맥락에서 항상 중요한 관심사다. 특히, 경제적 불평등을 이해하고 계량화하는 것은 사회경제적 담론의 핵심 주제다. 이러한 측면에서 로렌츠 곡선은 소득과 자산 분포를 시각적으로 나타내는 주요 분석 도구로 자리 잡고 있다.

여러 연속 모수적 분포 함수들이 로렌츠 곡선과 지니 계수(Gini coefficient)를 평가하는 방법으로 제안되었지만, 대부분은 총소득(gross income)을 중심으로 하며, 양수 또는 비음수 범위에서만 정의된다. 예로는 로그정규(log-normal), 와이블(Weibull), 감마(Gamma), Singh-Maddala (Singh and Maddala, 1975), Dagum (Dagum, 1977), GB2 (McDonald, 1984), dPLN (Reed, 2003), 및 κG (Clementi et al., 2007) 분포가 있다.

최근에는 총소득뿐만 아니라 순자산이나 처분가능소득의 분포에 대한 관심이 증가하고 있다. 하지만 이러한 변수들은 음수와 0 값을 포함하고 있어 기존의 총소득 분포 추정에 사용된 분포 함수들을 적용할 수 없다(Jäntti et al., 2015). 본 연구는 Johnson(1949a)의 $S_U$분포를 활용하여 순자산이나 소득의 분포를 추정할 것을 제안한다. $S_U$분포 함수는 실수 전체에서 정의되며, 다양한 비대칭과 첨도를 포괄할 수 있는 가장 유연한 모수적 분포 중 하나이다. 따라서 음수를 포함할 수 있는 순자산이나 처분가능소득의 다양한 분포 형태를 추정하는 데 적합하다. 우리는 $S_U$분포 함수에 대한 닫힌 형태의 로렌츠 곡선을 도출하였다. 예시로, 미국 소비자 금융조사(SCF) 데이터를 사용하여 순자산과 소득의 분포 및 로렌츠 곡선을 추정하였다.

## 2 Johnson $S_U$ 분포

$S_U$분포는 Johnson (1949a)의 획기적인 논문에서 처음 등장하였다. $S_U$변수 $X$는 다음과 같은 정규성 변환을 통해 생성된다.  

$$
\sinh^{-1} \left(\frac{X-m}{s}\right) = \lambda + \theta Z, \quad -\infty < X < \infty, \quad s > 0, \quad \theta > 0
$$

여기서 $Z$는 표준 정규 변수이다. $S_U$는 "무한 시스템(unbounded system)"을 의미하며, $X$의 범위가 무한임을 나타낸다. $S_U$의 확률 밀도 함수(PDF)는 다음과 같다.  

$$
f(x) = \frac{1}{\theta \sqrt{(x-m)^2 + s^2}} \phi\left(\theta^{-1} \left[\sinh^{-1} \left(\frac{x-m}{s}\right) - \lambda\right]\right)
$$

여기서 $\phi(\cdot)$는 표준 정규 변수의 PDF이다. 누적 분포 함수(CDF)는 다음과 같다.  

$$
F(x) = \Phi\left(\theta^{-1} \left[\sinh^{-1} \left(\frac{x-m}{s}\right) - \lambda\right]\right)
$$

여기서 $\Phi(\cdot)$는 표준 정규 변수의 CDF이다.

Johnson(1949a)은 $X$의 처음 네 가지 모멘트를 다음과 같이 제시하였다.
</br></br>

<div style="text-align: left; margin-left: 4em;">
$
\mu_1' \equiv \mu = m + s\omega^{1/2} \sinh(\lambda)
$
</br></br>
$
\mu_2 \equiv \sigma^2 = \frac{1}{2}s^2 (\omega - 1)(\omega \cosh(2\lambda) + 1)
$  
</br></br>
$
\mu_3 = \frac{1}{4}s^3 \omega^{1/2} (\omega - 1)^2 \left[\omega(\omega+2)\sinh(3\lambda) + 3\sinh(\lambda)\right]
$  
</br></br>
$
\mu_4 = \frac{1}{8}s^4 (\omega-1)^2 \left[\omega^2(\omega^4 + 2\omega^3 + 3\omega^2 - 3)\cosh(4\lambda) + 4\omega^2(\omega+2)\cosh(2\lambda) + 3(2\omega+1)\right]
$
</div>

여기서 $\omega = e^{\theta^2}$이다. $X$의 왜도(skewness)와 첨도(kurtosis)는 각각 $\mu_3 / \sigma^3$와 $\mu_4 / \sigma^4$이다.  

$S_U$분포는 매우 유연한 분포 함수로, 왜도와 초과 첨도의 가장 넓은 조합을 포착할 수 있다. 이 유연성 덕분에 $S_U$분포는 다양한 분야에서 응용되고 있다. 특히, 단변량 및 다변량 금융 수익률 모델링과 Value-at-Risk 추정에서 우수한 성과를 보였다(Choi and Nam, 2008 등).  

로렌츠 곡선은 자산 분포를 그래픽으로 나타내는 도구이다. $S_U$변수 $X$와 $p = F(x)$에 대해 로렌츠 곡선은 다음과 같다(부록 1 참조).  

$$
L(p) = \frac{1}{m + s e^{\theta^2 / 2} \sinh(\lambda)} \left[mp + \frac{1}{2} s e^{\theta^2 / 2} \left(e^{\lambda}\Phi(\Phi^{-1}(p) - \theta) - e^{-\lambda}\Phi(\Phi^{-1}(p) + \theta)\right)\right]
$$

$n$개의 관측치와 밀도 $f(x_i)$, $i = 1, \ldots, n$가 주어졌을 때, 우도 함수는 다음과 같다.  

$$ 
\ln L = \sum_{i=1}^n \ln f(x_i) = \sum_{i=1}^n \left(-\ln \theta - \frac{1}{2}\ln((x_i - m)^2 + s^2) - \frac{1}{2}\ln(2\pi) - \frac{1}{2}\left[\theta^{-1} \left(\sinh^{-1}\left(\frac{x_i - m}{s}\right) - \lambda\right)\right]^2\right)
$$

## 3 예제

2022년 SCF 데이터셋에서는 가계 재정에 특히 중요한 두 변수인 소득과 순자산에 집중합니다. 두 변수 모두 1,000 미국 달러 단위로 측정되며, 관측값의 수는 4,595 가구입니다. 표 1은 두 변수에 대한 주요 통계치를 제공합니다. 두 변수 모두 극단적인 양의 왜도와 과잉 첨도를 나타내며, 이는 SU 분포가 이들에 적합한 분포 함수가 될 수 있음을 시사합니다. 표의 마지막 행은 경험적 지니 계수를 제공하며, 순자산(0.830)의 불평등 수준이 소득 불평등(0.607)보다 상당히 높음을 나타냅니다.

**표 1. 순자산과 소득에 대한 주요 통계 (단위: 1만 달러)**

|          | 순자산    | 소득     |
|----------|-----------|----------|
| 평균     | 1,059.5   | 141.4    |
| 표준편차 | 8,085.5   | 728.0    |
| 왜도     | 92.8      | 166.8    |
| 첨도     | 14,517.1  | 63,910.6 |
| 지니 (×100) | 83.0      | 60.7     |

식 (1)에 대한 최대 우도 추정은 Python SciPy의 'optimize' 모듈과 제한 메모리 BFGS(L-BFGS-B) 알고리즘을 사용하여 수행되었습니다. 표 2에서 두 개의 모양 매개변수($\hat{\lambda}$와 $\hat{\theta}$)를 비교한 결과, 순자산이 소득보다 더 불평등하게 분포되어 있음을 알 수 있습니다. 또한, SU 모델을 통한 순자산의 지니 계수는 경험적 지니 계수보다 높게 추정되었고, 소득은 반대로 추정되었습니다.

**표 2. SU 분포 추정 결과**

|          | 순자산    | 소득     |
|:--------:|-----------|----------|
| 파라미터 |           |          |
| $\hat{m}$      | -4.354    | 7.615    |
| $\hat{s}$       | 14.876    | 10.955   |
| $\hat{\lambda}$      | 3.046     | 2.427    |
| $\hat{\theta}$      | 2.090     | 1.130    |
| 지니계수 ($\times$100) | 86.7      | 54.9     |

그림 1과 2의 왼쪽 패널에 있는 히스토그램에서 보듯이, 순자산과 소득 모두 극단적인 양의 왜도를 가진 분포를 나타냅니다. 그럼에도 불구하고, 추정된 SU 분포는 경험적 분포를 잘 나타냅니다. 표 1과 2의 오른쪽 패널에는 두 변수에 대한 경험적 및 추정된 로렌츠 곡선이 표시되어 있습니다. 순자산의 경우, SU 모델에 의한 로렌츠 곡선이 경험적 곡선보다 약간 낮게 위치하며, 소득의 경우 반대가 관찰됩니다. 결과적으로, 순자산에 대한 SU 모델의 지니 계수 추정값은 표 2에서 보듯 경험적 지니 계수보다 높고, 소득에 대해서는 그 반대입니다.

**그림 1. 순자산에 대해 SU 모델을 사용하여 추정된 분포와 로렌츠 곡선**

**그림 2. 소득에 대해 SU 모델을 사용하여 추정된 분포와 로렌츠 곡선**

SU 분포의 유용성과 유연성을 평가하기 위해, 우리는 소득 분포 추정에 가장 기본적인 분포 함수로 간주될 수 있는 로그정규 분포와 비교합니다. 그러나 로그정규 분포는 변수 값이 0보다 커야 하므로 순자산 변수에는 적용할 수 없습니다. 또한, 소득 데이터셋에 일부 0 값이 포함되어 있으며(전체 가구의 0.86%), 이는 로그정규 분포를 적용하기 전에 해결해야 합니다. 우리는 소득 값이 0인 경우를 작은 양의 값($10^{-5}$)으로 대체하여 추정을 진행했습니다.

그림 3에 그 결과가 나와 있습니다. 첫째, 그림의 왼쪽 패널에서 보여지는 로그정규 모델의 적합도를 SU 모델의 결과(그림 2)와 비교해 보면, 후자의 추정 성능이 더 우수한 것을 시각적으로 확인할 수 있습니다. 둘째, 로렌츠 곡선을 추정하는 측면에서, 앞서 보았듯이 SU 모델은 경험적 로렌츠 곡선에 비해 소득 불평등을 과소 추정하는 반면, 그림 3의 로그정규 모델은 소득 불평등을 과대 추정하는 경향이 있습니다. 로렌츠 곡선을 추정하는 데 있어 어느 모델이 더 잘 맞는지는 그림만으로는 명확하지 않지만, 추정된 지니 계수를 비교하면, 로그정규 모델(0.687)의 지니 계수가 경험적 지니 계수(0.607)와 비교해 더 큰 차이를 보이며, SU 모델(0.549)은 그보다 작은 차이를 보입니다.

**그림 3. 소득에 대해 로그정규 모델을 사용하여 추정된 분포와 로렌츠 곡선**


## 4 결론

자산이나 소득에 대한 모수적 분포 모델은 적은 수의 매개변수로 분포의 모든 특징을 포착할 수 있다는 장점이 있습니다. 또한, 이러한 모델은 설문 조사 마이크로데이터가 없고 그룹화된 데이터만 제공될 때도 분포 및 불평등/빈곤 지표를 추정하는 데 유용합니다. 다양한 모수적 분포가 소득 분포 추정에 사용되어 왔지만, 부채나 순자산처럼 음수와 0 값을 포함하는 변수에는 적용할 수 없습니다. 이러한 변수들의 분포를 추정할 때, SU 분포는 전체 실수 범위에서 정의되고 극단적인 왜도와 첨도를 잘 포착할 수 있기 때문에 가장 적합한 후보 중 하나로 고려될 수 있습니다. 
또한, SU 분포는 본질적으로 정규 분포의 변환이기 때문에 다변량 차원으로 쉽게 확장할 수 있는 장점이 있습니다. 예를 들어, 이변량 SU 분포를 사용하여 순자산과 소득의 결합 분포를 추정할 수 있습니다.

## 부록: $S_U$ 분포 로렌츠 곡선 도출

SU 변수 $X = m + s \cdot \sinh(\lambda + \theta Z)$를 확률 밀도 함수 (PDF) $f(x)$와 누적 분포 함수 (CDF) $F(x)$를 사용하여 고려합시다. 여기서 $p$는 인구 중에서 순자산이 $x$보다 낮은 사람들의 비율을 나타내며, 즉 $p = F(x)$입니다. 그럼 로렌츠 곡선 $L(p)$는 다음과 같이 정의됩니다 (Lubrano, 2017).

$$
L(p) = \frac{1}{E[X]} \int_{-\infty}^{x} t f(t) \, dt = \frac{1}{m + s e^{\theta^2/2} \sinh(\lambda)} \int_{-\infty}^{x} t f(t) \, dt \tag{A1}
$$

정규 랜덤 변수 $Y \sim N(\mu, \sigma)$가 $a \leq Y \leq b$일 조건에서, 그 순간 생성 함수(moment generating function)는 다음과 같습니다:

$$
E[e^{tY} \mid a \leq Y \leq b] = \frac{e^{t\mu + \frac{1}{2} t^2 \sigma^2}}{\Phi\left(\frac{b - \mu}{\sigma}\right) - \Phi\left(\frac{a - \mu}{\sigma}\right)} \left[ \Phi\left( \frac{b - \mu}{\sigma} - t\sigma \right) - \Phi\left( \frac{a - \mu}{\sigma} - t\sigma \right) \right] \tag{A2}
$$

SU 랜덤 변수 $X$가 $a \leq X \leq b$일 조건에서, $\sinh^{-1}\left(\frac{X - m}{s}\right) \equiv Y = \lambda + \theta Z$로 표현될 때, 첫 번째 제로에 대한 모멘트는 다음과 같습니다:

$
E[X \mid a \leq X \leq b] = E[m + s \cdot \sinh(Y) \mid a \leq m + s \cdot \sinh(Y) \leq b]
$
<div style="text-align: left; margin-left: 4em;">
$
= m + s E\left[\frac{1}{2} (e^Y - e^{-Y}) \mid \sinh^{-1}\left(\frac{a - m}{s}\right) \leq Y \leq \sinh^{-1}\left(\frac{b - m}{s}\right) \right]
$
$
= m + \frac{s}{2} \left\{ E\left[e^Y \mid \sinh^{-1}\left(\frac{a - m}{s}\right) \leq Y \leq \sinh^{-1}\left(\frac{b - m}{s}\right) \right] - E\left[e^{-Y} \mid \sinh^{-1}\left(\frac{a - m}{s}\right) \leq Y \leq \sinh^{-1}\left(\frac{b - m}{s}\right) \right] \right\}
$
$
= m + \frac{1}{2} s e^{\theta^2/2} \frac{ e^{\lambda} \left[\Phi(\beta - \theta) - \Phi(\alpha - \theta)\right] - e^{-\lambda} \left[\Phi(\beta + \theta) - \Phi(\alpha + \theta)\right] } { \Phi(\beta) - \Phi(\alpha) } \text{ (by A2)} 
$
</div>


여기서 $\alpha = \theta^{-1} \left[ \sinh^{-1}\left( \frac{a - m}{s} \right) - \lambda \right]$이고, $\beta = \theta^{-1} \left[ \sinh^{-1}\left( \frac{b - m}{s} \right) - \lambda \right]$입니다. $a = -\infty$와 $b = x$일 때, 위 식은 다음과 같이 변형됩니다:

$$
E[X | -\infty < X \leq x] = m + \frac{1}{2} s e^{\theta^2/2} \frac{ e^{\lambda} \Phi\left( \theta^{-1} \left[ \sinh^{-1}\left( \frac{x - m}{s} \right) - \lambda \right] - \theta \right) - e^{-\lambda} \Phi\left( \theta^{-1} \left[ \sinh^{-1}\left( \frac{x - m}{s} \right) - \lambda \right] + \theta \right) } {\Phi\left( \theta^{-1} \left[ \sinh^{-1}\left( \frac{x - m}{s} \right) - \lambda \right] \right)}
$$

따라서, $E[X | -\infty \leq X \leq x] = \frac{1}{F(x)} \int_{-\infty}^{x} t f(t) \, dt$이고, $F(x) \equiv p = \Phi\left( \theta^{-1} \left[ \sinh^{-1}\left( \frac{x - m}{s} \right) - \lambda \right] \right)$이므로, 우리는 다음을 얻습니다:

$$
\int_{-\infty}^{x} t f(t) \, dt = m p + \frac{1}{2} s e^{\theta^2/2} \left[ e^{\lambda} \Phi\left( \Phi^{-1}(p) - \theta \right) - e^{-\lambda} \Phi\left( \Phi^{-1}(p) + \theta \right) \right]
$$

따라서, (A1)은 다음과 같이 변형됩니다.

$$
L(p) = \frac{1}{m + s e^{\theta^2/2} \sinh(\lambda)} \left\{ m p + \frac{1}{2} s e^{\theta^2/2} \left[ e^{\lambda} \Phi\left( \Phi^{-1}(p) - \theta \right) - e^{-\lambda} \Phi\left( \Phi^{-1}(p) + \theta \right) \right] \right\}
$$
