### 다문화청소년 패널 데이터 만들기

원시(raw) 데이터를 사용해 분석을 시작하기를 원하는 독자들은 한국청소년정책연구원에서 데이터를 [다운로드](https://www.nypi.re.kr/archive/mps/program/examinDataCode/dataDwloadAgreeView?menuId=MENU00226) 받아야 한다. 1차년도부터 7차년도까지의 데이터를 다운받아 이를 패널 데이터로 만드는 작업부터 시작해야 한다. 아래 명령문이 그것에 해당한다.

    import pandas as pd
    import os
    import re

    # 폴더 경로
    base_folder = r"C:\1.1 다문화\Data\KCYPS2010 e4[EXCEL]"

    # 패널 데이터를 만들기 위한 반복문
    panel_data_list = []

    for i in range(1, 8):  # 1차년도부터 7차년도까지
        file = os.path.join(base_folder, f"KCYPS2010 e4w{i}.xlsx")

        # 파일 로딩 및 변수명에서 "w{연차}" 접미사 제거
        data = pd.read_excel(file).rename(
            columns=lambda x: re.sub(f"w{i}$", "", x) #$는 문자열의 끝을 나타냄
        )
    
        # 각 연차별 데이터프레임을 리스트에 추가
        panel_data_list.append(data)

    # 모든 연차의 패널 데이터를 하나로 결합하여 "panel_data" 생성
    panel_data = pd.concat(panel_data_list)

### 분석 대상 변수 만들기

KCYPS 2010의 조사 항목은 크게 세 가지 영역으로 구분된다. 첫 번째 영역은 성별, 가구소득, 부모학력 등 배경 변인이고, 두 번째 영역은 건강, 지적발달, 사회정서발달 등 개인 발달 영역에 속하는 항목이며, 마지막 영역은 가정환경, 친구관계, 교육환경 등 발달 환경 영역에 속하는 항목이다.

본 연구에서 분석하고자 하는 설문 항목은 학업성적이다. 이것은 “지난 1학기 동안 다른 친구들에 비해 얼마나 공부를 잘했다고 생각하느냐?”는 질문에 대해 학생 본인이 스스로 평가하는 문항이다. 각 과목별로 1(매우 못했다)부터 5(매우 잘 했다)까지 5점 척도로 평가되었다. 원래 설문조사에서는 1에서 5로 갈수록 성적이 낮은 방식으로 되어 있지만, 우리는 회귀모형 계수의 해석 편의를 위해 이것을 반대 방향(즉, 값이 높을수록 성적이 좋은 방식)으로 리코딩했다. 우리는 전체 과목 중 국어 수학 영어 과학 사회 등 5개 과목의 평점을 평균하여 단일의 성적(`score`) 변수를 만들었으며, 이를 분석의 목표변수로 삼았다.

우리는 주어진 아동·청소년패널조사에서 가급적 많은 문항들을 설명변수로 사용하고자 했다. 초1・초4・중1의 세 가지 패널조사에서 공통적으로 조사된 항목 중에서 학업성적과 관련성을 가진다고 판단되는 대부분의 설문항목들을 설명변수에 포함시켰다. `male`(남학생 더미변수), `income`(가구 연간 소득), `fa_edu`(아버지 학력), `mo_edu`(어머니 학력), `par_satis`(보호자 삶의 만족도), `health`(본인 건강상태), `aft_sch`(학원・과외시간), `aft_sch1`(방과후교육 경험 더미변수), `presch`(초등 전 사교육 경험 더미변수), `care`(부모의 관심), `abuse`(부모의 학대), `sch_study`(학교 학습활동), `friend`(교우 관계), `teacher`(교사 관계) 등이다. 각 변수에 대한 자세한 설명이 표 1에 나와 있다.

설문항목을 변수로 만드는 과정에서 리코딩 등의 작업이 가미되기도 했다. 가령 조사대상 학생 본인의 건강상태를 조사한 `health` 변수의 경우, 원래의 조사에서는 1의 값이 가장 건강하고 4의 값이 가장 건강하지 않은 것을 의미하나, 회귀분석 계수 해석의 편의를 위해 이를 반대로 리코딩하여 4의 값이 가장 건강한 것으로 바꾸었다. 그리고 `aft_sch`(학원・과외시간)의 경우, 원래는 등교일과 비등교일에 대해 각각 조사됐으나, 우리는 이를 등교일과 비등교일을 2대 1의 비율로 가중평균하여 하나의 값으로 만들었다.

**`mc` 변수**

```python
# ID별로 그룹화
grouped = panel_data.groupby('ID')

# 각 ID에 대해 최빈값 계산
result = {}
for name, group in grouped:
    # 결측값이 아닌 데이터만 고려
    non_missing_data = group['mcd'].dropna()
    if non_missing_data.empty:
        result[name] = None  # 모든 값이 결측인 경우
    else:
        counts = non_missing_data.value_counts()
        if len(counts) > 1 and counts.iloc[0] == counts.iloc[1]:
            result[name] = 1  # 0과 1의 개수가 동일한 경우 1 선택
        else:
            result[name] = counts.idxmax()  # 가장 많이 나온 값

# 새로운 컬럼 mc을 생성하여 ID별로 계산된 값 매핑
panel_data['mc'] = panel_data['ID'].map(result)

# 데이터프레임 출력
panel_data['mc']
```

**부모 교육수준이 여러 값을 갖는 경우 최빈값으로 통일: `fa_edu` & `mo_edu`**

```python
import pandas as pd
import numpy as np

# ID별로 그룹화
grouped = panel_data.groupby('ID')

# 각 ID에 대해 최빈값 계산
def calculate_mode(group):
    # fa_ed1의 최빈값 계산
    mode_value = group['fa_ed1'].mode()
    if not mode_value.empty:
        return mode_value.iloc[0] # iloc[0]는 첫 번째(가장 크기가 작은) 최빈값
    else:
        return np.nan  # 모든 값이 결측인 경우 NaN 반환

# 각 ID에 대한 최빈값을 가진 시리즈 생성
fa_edu = grouped.apply(calculate_mode)

# 새로운 컬럼 fa_edu를 생성하여 ID별로 계산된 최빈값 매핑
panel_data['fa_edu'] = panel_data['ID'].map(fa_edu)

# 결과를 확인
print(panel_data[['ID', 'fa_ed1', 'fa_edu']])
```

```python
import pandas as pd
import numpy as np

# ID별로 그룹화
grouped = panel_data.groupby('ID')

# 각 ID에 대해 최빈값 계산
def calculate_mode(group):
    # fa_ed1의 최빈값 계산
    mode_value = group['mo_ed1'].mode()
    if not mode_value.empty:
        return mode_value.iloc[0] # iloc[0]는 첫 번째(가장 크기가 작은) 최빈값
    else:
        return np.nan  # 모든 값이 결측인 경우 NaN 반환

# 각 ID에 대한 최빈값을 가진 시리즈 생성
mo_edu = grouped.apply(calculate_mode)

# 새로운 컬럼 mo_edu를 생성하여 ID별로 계산된 최빈값 매핑
panel_data['mo_edu'] = panel_data['ID'].map(mo_edu)

# 결과를 확인하려면 다음을 출력하십시오.
print(panel_data[['ID', 'mo_ed1', 'mo_edu']])
```

## 3 패널 분석

### 추정 모형

추정 모형은 다음과 같다.

$$
y_{it} = \beta_0 + \beta_1 m_i + \beta_2 x_{it} + \beta_3 x_{it} m_i + \beta_4 d_t + u_i + e_{it} \tag{1}
$$

여기에서 $i$는 패널에 속한 어떤 개인을 나타내고, $t$는 개인의 특성이 관찰된 어떤 시점을 의미하며, $y_{it}$는 시점 $t$에서 개인 $i$의 학업성적을 나타낸다. 우변 첫 번째 항인 $m_i$는 다문화 더미변수로서 다문화 학생이면 1, 그렇지 않으면 0의 값을 가지며, 시점에 상관없이 동일한 값을 갖는다. 소수이기는 하지만, 주어진 조사기간 중 다문화가정 여부에 대한 응답이 바뀌는 경우도 발견되는데, 다수결 원칙을 사용해 이를 하나의 값으로 통일시켰다. 

$x_{it}$는 학업성적에 영향을 미치는 설명변수로서 가구의 소득 수준이 대표적 예이다. $x_{it}$에 아래첨자 $t$가 붙어 있지만, 남녀 성별 변수와 같은 시간불변(time-invariant) 변수도 설명변수가 될 수 있다. $d_t$는 연도별 특성을 포착하기 위한 시간 더미변수로서, 가령 $d_{2012}$는 연도가 2012년이면 1의 값을 갖고, 그 외에는 0의 값을 갖는다. $u_i$는 개인별 이질성(heterogeneity)을 나타내는 것으로서 확률효과(random effect) 모형에서는 이것을 확률변수로 간주한다. 

KCYPS 데이터의 경우 데이터의 시점($t$)의 개수가 다섯 개(중1 패널) 또는 여섯 개(초1・초4 패널)에 불과한 데 비해, 개인($i$)의 숫자는 이보다 훨씬 많기 때문에 $u_i$를 고정된 값으로 간주하는 고정효과(fixed effect) 모형에 비해 확률효과 모형이 추정에서 더 효율적이라고 판단했다. 또한 위에 제시된 모형에서 가장 중요한 변수인 $m_i$가 시간불변 변수라서 고정효과 모형을 사용할 경우 $m_i$의 계수를 추정할 수 없다는 문제도 있다.

위 모형에서 다문화 더미변수 $m_i$의 계수인 $\beta_1$은 본 연구에서 가장 중요한 계수로서 만약 이것이 플러스(마이너스) 부호이면, 다른 조건($x_{it}$ 및 $d_t$)이 동일할 때 다문화 학생의 학업성적이 비다문화 학생에 비해 더 높은(낮은) 경향이 있다는 것을 의미한다. 한 가지 유의할 것은 $\beta_1$의 크기가 두 그룹 학업성적의 평균적인 차이는 아니라는 점이다. 이는 위 식의 우변에 상호작용항($x_{it} m_i$)이 들어 있기 때문이다. 따라서 다른 조건이 동일한 상태에서 $m_i$가 0일 때와 1일 때의 평균적인 차이는 $\beta_1$에다 $\beta_3 x_{it}$를 더한 값이다.

위 식 1에서 상호작용항 $x_{it} m_i$는 설명변수 $x_{it}$가 학업성적에 영향을 미치는 효과에 있어서 다문화 학생과 비다문화 학생 간에 어떤 차이가 있는지를 포착하는 역할을 한다. 다른 조건이 동일할 때, $x_{it}$가 학업성적 $y_{it}$에 미치는 평균적인 효과는 비다문화 학생의 경우에는 $\beta_2$이지만, 다문화 학생의 경우에는 그것에다 상호작용항의 계수를 더한 $\beta_2 + \beta_3$가 된다. 따라서 상호작용항의 계수 $\beta_3$만큼 다문화 학생과 비다문화 학생 간에 소위 한계효과(marginal effect)의 차이가 있는 것이다.


### 추정 결과

이상에서 설명한 패널 회귀모형을 확률효과 가정 하에서 추정한 결과가 표 3에 나와 있다. 표의 위쪽에는 개별 설명변수에 대한 추정 결과가 나와 있고, 중간에는 상호작용항에 대한 추정 결과, 그리고 표의 하단에는 시간 더미변수에 대한 추정 결과가 나와 있다. 맨 아래쪽의 시간 더미변수에 대한 결과부터 보면, 세 개의 패널 데이터 모두 2011년을 기준 연도로 삼아 추정했다는 것을 알 수 있다. 또한 중1 패널의 경우에는 2015년 고3까지의 데이터만 사용했기 때문에 $d_{2016}$ 변수가 모형에 들어있지 않다. 

초1 패널의 경우 시간 더미변수의 추정 계수 부호가 모두 마이너스이고 통계적으로 매우 유의하다. 또한 각 계수의 추정값을 보면, 다른 조건이 동일할 때 학년이 높아짐에 따라 학생들의 성적이 낮아지는 경향이 있다는 것을 말해준다. 초4 패널의 시간 더미변수 추정 결과도 초1 패널과 전체적으로는 유사하지만, 학업성적의 하락 경향이 초1만큼 뚜렷하지는 않다. 마지막으로 중1 패널의 시간 더미변수 추정 결과는 초1・초4와는 다른데, 중2(2011년)에서 중3(2012년), 고1(2013년)까지는 성적이 낮아지지만, 마지막 고3(2015년)의 성적은 중2 성적보다도 높아진다. 이러한 회귀분석 결과는 앞에서 그림 2에서 그림을 통해 살펴본 것과 유사한 결과이다.

이제 표 3의 맨 위쪽에 나와 있는 개별 변수들에 대한 추정 계수 부호와 통계적 유의성을 살펴보자. 본 연구에서 가장 중요한 변수인 다문화 더미변수(`m`)의 경우, 초1・초4 패널의 경우 계수 부호가 마이너스이다. 이는 다른 조건(소득수준, 부모님 학력, 학원・과외시간 등)이 동일할 때 다문화 학생의 성적이 더 낮은 경향이 있음을 의미한다. 그러나 이에 대한 통계적 유의성이 그다지 강하지 않은데, 둘 다 1% 수준에서 유의하지는 않고, 초1 패널은 5% 수준에서 유의하고, 초4 패널은 10% 수준에서 유의하다.

결과적으로 본 연구의 핵심 이슈, 즉 제반 요인들을 통제한 상태에서 다문화 학생의 학업성적이 상대적으로 뒤지는지 여부에 있어서 우리는 자신있게 그렇다고 답하기 어려운 결과를 얻었다. 초1・초4 패널의 경우, 다문화 학생의 학업성적이 상대적으로 뒤지는 것을 의미하는 부호가 나왔지만 통계적 유의성이 그다지 강하지 않고, 중1 패널의 경우에는 오히려 부호가 반대라는 강력한 증거가 나왔기 때문이다. 요컨대, 다문화 배경이 학업성적에 미치는 효과에 있어서 우리는 통계적 강건성을 지닌 결과를 얻지 못했다고 말할 수 있다.

표 3에서 다문화 더미변수 이외의 다른 설명변수들을 보면, 우리가 선택한 대부분의 변수가 통계적 유의성이 높게 나온 것을 알 수 있다. 또한 많은 경우에 있어서 추정된 계수의 부호가 예상과 일치할 뿐 아니라 초1・초4・중1 세 개의 데이터세트 간에 부호가 상충되지 않고 일관되게 나왔다. 가령 `income` 변수의 경우 세 개의 패널 모두 추정 계수가 통계적으로 유의하며, 플러스 값으로 나왔다. 가구 소득이 높은 환경에 속한 학생일수록 학업성적이 높은 경향이 있다는 것이다. 부모 학력(`fa_edu`, `mo_edu`)의 경우, 부와 모의 학력 모두 통계적으로 유의하며, 부모의 학력이 높을수록 자녀의 성적이 높은 경향이 있는 것으로 나왔다. 

한편, 배경변인 중 성별(`male`)의 경우에는 남학생이 여학생보다 학업성적이 더 높은(정확히 표현하면 자신의 학업성적을 더 높이 평가하는) 경향이 있지만, 초4・중1 패널에 있어서는 통계적 유의성이 떨어진다. 보호자 삶의 만족도(`par_satis`)의 경우에도 통계적 유의성이 떨어지는 바, 초4 패널을 제외하고 초1・중1 패널에서는 통계적으로 유의하지 않다.

한편, 개인발달과 관련된 설명변수 중 학생 본인 건강상태(`health`)의 경우, 건강하다고 느낄수록 학업성적이 높은 경향이 있으며, 이는 통계적으로 유의하다. 학원・과외시간(`aft_sch`), 방과후교육 경험(`aft_sch1`), 초등 전 사교육 경험(`presch`) 등도 추정 계수의 부호가 예상대로 플러스이고 통계적 유의성이 높다. 단, 예외적으로 중1 패널의 경우, `aft_sch1` 변수가 유의하지 않은 것을 알 수 있다.

발달환경과 관련된 설명변수들을 보면, 모두 다 학업성적에 미치는 효과가 예상대로 나왔으며, 일부 통계적 유의성이 결여된 경우도 있지만 전체적으로는 통계적 유의성이 높게 나왔다. 즉, 부모의 관심(care)이 높을수록, 부모의 학대(abuse)가 적을수록, 학교 학습활동(sch_study)이 우수할수록, 그리고 교우 관계(friend) 및 교사 관계(teacher)가 좋을수록 학업성적이 좋은 경향이 있다.

표 3의 중간 부분에 나와 있는 상호작용항에 대한 추정 결과를 보면, 개별 변수들에 대한 결과와는 반대이다. 즉, 대부분의 항이 통계적으로 유의하지 않다. 통계적 유의성을 가진 항도 드물게 찾아볼 수 있지만(가령, 초4 패널의 `sch_study・m`), 초1・초4・중1 패널 데이터 전체에 걸쳐 일관되게 유의성을 지닌 항은 하나도 없다. 

우리가 사용한 설명변수들이 개별적으로는 세 가지 모든 패널에 대해 일관되게 유의성을 가진 경우가 많은 반면, 이들 설명변수와 다문화 더미변수를 결합시켜 만든 상호작용항들은 대부분 유의성을 보이지 않다는 점이 중요하다. 이것은 우리가 사용한 설명변수들이 학업성적에 유의한 영향을 미치지만, 그 영향력에 있어서 다문화 학생과 비다문화 학생 간에 유의한 차이가 없다는 것을 의미한다. 달리 표현하면, 이들 설명변수와 학업성적의 관계가 다문화 배경 여부에 따라 달라지지는 않는다는 것이다. 가령 가정의 경제수준은 자녀의 학업성적과 플러스 관계가 있지만, 이들 두 변수 간의 관계가 다문화 학생이라고 해서 더 커진다거나 또는 작아진다고 볼 수 있는 통계적 증거가 없다는 것이다. 이러한 해석은 학원・과외시간이나 부모의 관심, 교우관계 등 모든 변수에 해당된다.

### 합동(pooled) OLS 모형

```python
import pandas as pd
from linearmodels.panel import PooledOLS, PanelOLS, RandomEffects, compare
from statsmodels.api import add_constant

# ID와 WAVE를 인덱스로 설정
data = df.set_index(['ID', 'WAVE'])

# 상호작용 항
data['male*mc'] = data['male'] * data['mc']
data['income*mc'] = data['income'] * data['mc']
data['fa_edu*mc'] = data['fa_edu'] * data['mc']
data['mo_edu*mc'] = data['mo_edu'] * data['mc']
data['par_satis*mc'] = data['par_satis'] * data['mc']

data['health*mc'] = data['health'] * data['mc']
data['aft_sch*mc'] = data['aft_sch'] * data['mc']
data['aft_sch1*mc'] = data['aft_sch1'] * data['mc']

data['care*mc'] = data['care'] * data['mc']
data['abuse*mc'] = data['abuse'] * data['mc']
data['sch_study*mc'] = data['sch_study'] * data['mc']
data['friend*mc'] = data['friend'] * data['mc']
data['teacher*mc'] = data['teacher'] * data['mc']

# X 변수(상수항, 시간 더미 및 상호작용 항)
X_with_const = add_constant(
    data[['mc',
          'male', 'income', 'fa_edu', 'mo_edu', 'par_satis',
          'health', 'aft_sch', 'aft_sch1', 'care', 'abuse',
          'sch_study', 'friend', 'teacher',
          'male*mc', 'income*mc', 'fa_edu*mc', 'mo_edu*mc', 'par_satis*mc',
          'health*mc', 'aft_sch*mc', 'aft_sch1*mc', 'care*mc', 'abuse*mc',
          'sch_study*mc', 'friend*mc', 'teacher*mc',
          'wave3', 'wave4', 'wave5', 'wave6', 'wave7']]
)

# Pooled OLS 모델을 사용하여 회귀 모델을 실행
mod = PooledOLS(data['score'], X_with_const)
pooled_results = mod.fit(cov_type='clustered', cluster_entity=True)

# 결과 요약을 출력
print(pooled_results)
```

## 부록 1: 패널 데이터 추정 모형 비교

출처: ChatGPT

---

다음은 주어진 패널 회귀식에서 합동(Pooled) OLS, 고정효과(Fixed Effects, FE), 확률효과(Random Effects, RE) 모형의 차이점을 알기 쉽게 설명한 내용입니다.  

**주어진 회귀식**

$$
y_{it} = \beta_0 + \beta_1 m_i + \beta_2 x_{it} + \beta_3 x_{it} m_i + \beta_4 d_t + u_i + e_{it}
$$

- $y_{it}$: 개체 $i$의 시간 $t$에서의 종속 변수  
- $m_i$: 개체 $i$의 고정 특성 (다문화 여부 등, 시간에 따라 변하지 않는 변수)  
- $x_{it}$: 시간에 따라 변하는 설명 변수  
- $x_{it} m_i$: 상호작용항  
- $d_t$: 시간 고정 효과를 나타내는 더미 변수  
- $u_i$: 개체 $i$의 시간 불변 고유 효과 (미관측 이질성)  
- $e_{it}$: 잔차, 시간과 개체에 따라 변하는 오차항  

### 합동 OLS 모형

특징:  
- $u_i$를 무시하고 모든 데이터(모든 $i$와 $t$)를 단일 회귀모형으로 분석.  
- 개체 간 이질성(차이점)을 고려하지 않음.  

장점:  
- 단순하고 계산이 쉽다.  

단점:  
- $u_i$가 $x_{it}$와 상관이 있을 경우(내생성 문제), 추정치가 편향되고 일관되지 않음.  
- 개체 간 고유한 차이를 무시하기 때문에 설명력이 낮아질 가능성.  

비유: 모든 개체를 "동일한 학생"이라고 가정하고, 개체 간 차이를 고려하지 않는 학급 평균 계산.  

### 고정효과 모형

특징:  
- $u_i$를 포함하되, 이를 제거하기 위해 개체마다 평균을 빼는 방식으로 시간 불변 효과 제거.  
- $u_i$와 $x_{it}$가 상관이 있어도 편향 문제를 해결할 수 있음.  

장점:  
- 개체 간의 고유한 차이를 통제하여 $x_{it}$의 순수한 효과를 추정 가능.  
- $u_i$가 $x_{it}$와 상관되어 있을 때 유리.  

단점:  
- $m_i$처럼 시간에 따라 변하지 않는 설명변수의 효과(예: 다문화 여부)는 추정할 수 없음.  
- 데이터에서 시간에 따른 변화를 활용하므로, 시간 변동성이 적은 데이터에서는 비효율적.  

비유: 같은 학생이 매년 시험 점수가 얼마나 바뀌었는지 비교하여 학생 간 차이는 무시하고, 점수 변화만 분석.  

### 확률효과 모형

특징:  
- $u_i$를 확률적으로 분리된 오차항으로 처리($u_i \sim N(0, \sigma^2)$).  
- $u_i$와 $x_{it}$가 상관이 없다고 가정.  

장점:  
- $m_i$처럼 시간에 따라 변하지 않는 변수의 효과를 추정 가능.  
- 효율적이고 계산이 빠름.  

단점:  
- $u_i$와 $x_{it}$가 상관이 있다면(내생성 문제) 추정치가 편향되고 일관되지 않음.  

비유: 학생마다 고유한 특성은 무작위로 존재한다고 가정하고, 전체 집단 평균에 기반해 분석.  

**모형 선택 기준**

(1) $u_i$와 $x_{it}$의 상관성 여부  
   - 상관이 있으면 고정효과(FE)를 사용.  
   - 상관이 없으면 확률효과(RE)를 사용.

(2) 시간에 따라 변하지 않는 설명변수(예: $m_i$)의 효과를 추정하고 싶은지 여부  
   - 추정하고 싶다면 확률효과(RE) 선택.  
   - 필요 없다면 고정효과(FE) 선택.

(3) 하우스만 검정(Hausman Test)  
   - FE와 RE 중 적합한 모형을 선택하는 데 사용.  
   - 귀무가설: RE가 적합하다.  

**결론적으로**,  

- 데이터의 특성에 따라 모형을 선택해야 하며, FE는 개체 간 차이를 통제하는 데 유리하고, RE는 시간 불변 변수를 포함하는 분석에서 유리합니다.  
- 합동 OLS는 단순하지만, 현실적으로 $u_i$를 무시할 때 발생하는 문제를 고려해야 합니다.
