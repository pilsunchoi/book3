## 3 성적 변화 추이 그리기

### 결과 설명

그림 2는 다문화 학생과 비다문화 학생 두 그룹의 성적 변화를 초1, 초4, 중1 패널 모두에 대해 하나의 평면에 겹쳐 그린 것이다. 비다문화 학생의 성적은 실선으로 표시돼 있고, 다문화 학생은 점선으로 표시돼 있는데, 해당 패널에 속한 학생들의 성적을 평균한 값을 표시한 것이다. 패널들 간에 학년이 일부 겹치는 구간이 있고, 그 구간에서는 서로 다른 패널의 평균 성적이 (당연히) 똑같지 않다. 서로 다른 패널의 성적을 연결하여 그린 것이기는 하지만, 이 그림을 통해 학생들의 성적이 학년에 따라 어떤 변화를 보이는지 전체적인 트렌드를 파악할 수 있다.

<center> <b>그림 2.</b> 다문화・비다문화 학생 그룹의 성적 변화 추이 </center>

![성적 변화 추이](https://raw.githubusercontent.com/pilsunchoi/images3/main/1-2.png)

그림에서 가장 눈에 띄는 것은 학생들의 성적이 초등학교, 중학교, 그리고 고등학교 1학년 때까지는 점차 하락하는 경향을 보인다는 점이다. 이런 경향은 다문화・비다문화 학생 모두에게 공통적으로 나타나며, 특히 초등학교에서 중학교로 막 진학한 중1 시점에서 성적 하락폭이 매우 큰 것을 확인할 수 있다. 그러나 일단 고등학교에 진학한 이후에는 성적이 오히려 상승하는 경향을 보이는데, 다문화・비다문화 학생 모두 고1에서 고3으로 갈수록 성적이 점차 더 높아진다.

다문화・비다문화 학생 간의 학업성적 격차에 초점을 맞춰보면, 거의 모든 학년에서 다문화 학생의 성적(점선)이 비다문화 학생의 성적(실선)에 비해 더 낮은 것을 수 있다. 유일한 예외가 초1 패널의 초6년 성적으로서 두 그룹 간 성적이 거의 동일하게 나온 것을 제외하고는 모든 학년에서 다문화 그룹의 성적이 더 낮은 것으로 조사됐다.

### 평균 성적: 다문화학생 제외

```python
# 연차별 score 변수의 평균값 계산
score_e1 = df_e1.groupby('WAVE')['score'].mean()
score_e1 = df_e1[df_e1['mc'] == 0].groupby('WAVE')['score'].mean()
# 새로운 인덱스 값 설정
new_index = ['초2', '초3', '초4', '초5', '초6', '중1']
# 인덱스를 새로운 값으로 변경
score_e1.index = new_index

print(score_e1)
```

```python
# 연차별 score 변수의 평균값 계산
score_e4 = df_e4.groupby('WAVE')['score'].mean()
score_e4 = df_e4[df_e4['mc'] == 0].groupby('WAVE')['score'].mean()
# 새로운 인덱스 값 설정
new_index = ['초5', '초6', '중1', '중2', '중3', '고1']
# 인덱스를 새로운 값으로 변경
score_e4.index = new_index

print(score_e4)
```

```python
# 연차별 score 변수의 평균값 계산
score_m1 = df_m1.groupby('WAVE')['score'].mean()
score_m1 = df_m1[df_m1['mc'] == 0].groupby('WAVE')['score'].mean()
# 새로운 인덱스 값 설정
new_index = ['중2', '중3', '고1', '고2', '고3']
# 인덱스를 새로운 값으로 변경
score_m1.index = new_index

print(score_m1)
```

```python
# 비다문화 그룹 시리즈를 데이터프레임으로 결합
combined_df = pd.concat([score_e1, score_e4, score_m1], axis=1)
# 비다문화 그룹 컬럼 이름 설정
combined_df.columns = ['초1 패널', '초4 패널', '중1 패널']

combined_df
```

### 평균 성적: 다문화학생

```python
# 연차별 score 변수의 평균값 계산
mc_score_e1 = df_e1.groupby('WAVE')['score'].mean()
mc_score_e1 = df_e1[df_e1['mc'] == 1].groupby('WAVE')['score'].mean()
# 새로운 인덱스 값 설정
new_index = ['초2', '초3', '초4', '초5', '초6', '중1']
# 인덱스를 새로운 값으로 변경
mc_score_e1.index = new_index

print(mc_score_e1)
```

```python
# 연차별 score 변수의 평균값 계산
mc_score_e4 = df_e4.groupby('WAVE')['score'].mean()
mc_score_e4 = df_e4[df_e4['mc'] == 1].groupby('WAVE')['score'].mean()
# 새로운 인덱스 값 설정
new_index = ['초5', '초6', '중1', '중2', '중3', '고1']
# 인덱스를 새로운 값으로 변경
mc_score_e4.index = new_index

print(mc_score_e4)
```

```python
# 연차별 score 변수의 평균값 계산
mc_score_m1 = df_m1.groupby('WAVE')['score'].mean()
mc_score_m1 = df_m1[df_m1['mc'] == 1].groupby('WAVE')['score'].mean()
# 새로운 인덱스 값 설정
new_index = ['중2', '중3', '고1', '고2', '고3']
# 인덱스를 새로운 값으로 변경
mc_score_m1.index = new_index

print(mc_score_m1)
```

```python
# 다문화 그룹 시리즈를 데이터프레임으로 결합
mc_combined_df = pd.concat([mc_score_e1, mc_score_e4, mc_score_m1], axis=1)
# 다문화 그룹 컬럼 이름 설정
mc_combined_df.columns = ['초1 패널', '초4 패널', '중1 패널']

mc_combined_df
```

### 비다문화 & 다문화

```python
# 컬럼에 MultiIndex 설정
combined_df.columns = pd.MultiIndex.from_product([['비다문화'], combined_df.columns])
mc_combined_df.columns = pd.MultiIndex.from_product([['다문화'], mc_combined_df.columns])

# 두 데이터프레임을 하나로 결합
final_df = pd.concat([combined_df, mc_combined_df], axis=1)

# 결과 출력
print(final_df)
```

## 4 두 그룹의 평균 비교

본 연구에서 우리의 관심은 이와 같은 학업성적 차이가 대상 학생을 둘러싸고 있는 여러 배경 및 발달 변인을 동일하게 통제한 상태에서도 여전히 존재하는지 여부이다. 그것을 패널 회귀분석 모형으로 추정하지만, 간단히 두 그룹의 평균 성적 차이에 대한 통계적 유의성을 $t$-검정으로 추정해볼 수 있다. 

우리의 데이터가 패널 자료라는 점을 감안하지 않고, 각 패널에 대해 모든 시점 별로 $t$-검정을 실시한 결과, 1% 유의수준에서 차이가 나는 경우는 총 17개 시점에 대한 검정에서 단 하나도 발견되지 않았다. 유의수준을 5%로 할 경우, 초1・초4 패널 각 6개 시점 중 단 1개씩에서만 평균 성적의 차이가 유의했고, 중1 패널의 경우 5개 시점 중 2개 시점에서 평균값이 유의하게 달랐다. 

이는 그림 2에서 보는 것과 같은 학업성적 차이를 놓고 다문화 학생의 학업성적이 비다문화 학생에 비해 평균적으로 더 낮다고 단정지어서는 안 된다는 것을 의미한다. 물론 이 결과는 주어진 데이터가 패널 데이터라는 사실을 감안하지 않은 분석이라는 점에서 한계를 지닌다.

### 초1 패널 평균차이에 대한 $t$ 검정

```python
import scipy.stats as stats

# 결과를 저장할 리스트
t_test_results = []

# 각 WAVE에서 다문화학생과 비다문화학생의 평균값 차이 검정
for wave in df_e1['WAVE'].unique():
    # 비다문화학생 score
    non_mc_scores = df_e1[(df_e1['mc'] == 0) & (df_e1['WAVE'] == wave)]['score'].dropna()
    # 다문화학생 score
    mc_scores = df_e1[(df_e1['mc'] == 1) & (df_e1['WAVE'] == wave)]['score'].dropna()
    
    # t-검정 실행 (동분산 가정 여부는 데이터에 따라 결정)
    t_stat, p_value = stats.ttest_ind(non_mc_scores, mc_scores, equal_var=False)
    
    # 결과 저장
    t_test_results.append({
        'WAVE': wave,
        't-statistic': t_stat,
        'p-value': p_value,
        'non_mc_mean': non_mc_scores.mean(),
        'mc_mean': mc_scores.mean(),
        'non_mc_count': len(non_mc_scores),
        'mc_count': len(mc_scores)
    })

# 결과를 데이터프레임으로 변환하여 확인
t_test_df = pd.DataFrame(t_test_results)

# 새로운 인덱스 값 설정
new_index = ['초2', '초3', '초4', '초5', '초6', '중1']
# 인덱스를 새로운 값으로 변경
t_test_df.index = new_index

print(t_test_df)
```
